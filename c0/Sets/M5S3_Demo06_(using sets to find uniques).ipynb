Find all the unique elements in a list

```python
my_list = [1, 2, 3, 1, 1, 1, 3, 2, 4, 5, 6, 7]
```

This can be achieved using a loop

```python
unique_list = []  # stores unique elements in the list

for ele in my_list:  # iterate through the list
    if ele not in unique_list:  # only append an element to the new list if it isn't already there
        unique_list.append(ele)

print(unique_list)
```

A quicker way to do this would be to simply convert the list to a set and back

```python
list(set(my_list))
```

How can you eliminate all the duplicate characters in a string? Create a new string with the remaining characters.

```python
my_str = 'Hello world, I am learning Python!'
```

One way would be to use a loop

```python
unique_str = ''  # stores unique characters in the string

for ch in my_str:  # iterate through the string
    if ch not in unique_str:  # only concatenate the character to the new string if it isn't already in the string
        unique_str += ch

print(unique_str)
```

A quicker way would be to simply split the string to a list, convert the list to a set, and then make a new string using `.join()`

```python
''.join(set(my_str))
```

Note how the order of the characters may have changed. Sets do not guarantee an order in Python.

Now, given three lists, can you find the unique elements across all three lists?

```python
lst1 = [1, 2, 3, 4, 5]
lst2 = [2, 3, 4, 5, 6, 7]
lst3 = [5, 7, 9, 11, 13]
```

A quick way to do this would be to simply convert everything to a set and run a union

```python
set(lst1) | set(lst2) | set(lst3)
```
