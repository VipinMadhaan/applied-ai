Given a nested tuple containing pairs, can you sort the tuple by the second value of the pair?

```python
pairs = (('a', 4), ('b', 2), ('c', 3), ('d', 1))
```

```python
sorted_list = sorted(pairs, key = lambda x: x[1])  # access the second element for each pair while sorting
print(sorted_list)  # sorted() outputs a list
```

```python
print(tuple(sorted_list))  # converting the list back to a tuple
```

Given a nested tuple containing three pairs, can you make a new nested tuple containing two triplets such that the first element of each pair lies in the first triplet while the second element of each of each pair lies in the second triplet? For instance, the input `((1, 'a'), (2, 'b'), (3, 'c'))` should give `((1, 2, 3), ('a', 'b', 'c'))`.

**Tip:** Use the fact that a function outputs tuples when returning multiple values to your advantage.

```python
my_pairs = ((1, 'a'), (2, 'b'), (3, 'c'))
```

```python
def make_triplets(pairs):
    first_triplet = tuple([x[0] for x in pairs])  # make a list of every first element in the pairs and turn it into a tuple
    second_triplet = tuple([x[1] for x in pairs])  # make a list of every second element in the pairs and turn it into a tuple

    return first_triplet, second_triplet
```

```python
make_triplets(my_pairs)
```

Imagine a chessboard with coordinates corresponding to each square, ranging from `(0, 0)` to `(7, 7)`. Your piece is given a starting position `(x, y)` and a list of movements `(dx, dy)` where `dx` represents the movement on the x-axis, and `dy` represents the movement on the y-axis. Find the final position of your piece as a pair of coordinates. A piece stops moving if:

*   It runs out of movements in the list.
*   The move would move the piece out-of-bounds.

```python
starting_position = (5, 4)
movements = [(1, 2), (-1, -3), (2, 0), (0, 3), (1, 0)]
```

```python
def move(start, moves):
    x, y = start  # unpack the coordinates in the tuple

    i = 0  # iterator moving through the movements list
    while i < len(moves):
        dx, dy = moves[i]  # unpack the movement coordinates in the tuple

        if x + dx < 0 or x + dx > 7 or y + dy < 0 or y + dy > 7:  # check if the new coordinates are out-of-bounds and break if True
            break

        # If the moves are not out-of-bounds, you can update the coordinates of the piece
        x += dx  # update the x-coordinate
        y += dy  # update the y-coordinate

        i += 1  # update the iterator

    return x, y
```

```python
move(starting_position, movements)
```
