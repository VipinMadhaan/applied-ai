We can access a dictionary by a key to retrieve the value associated with that key. One way to do this is to use subscript notation.

```python
my_dict = {'student': 'Ashok Kummar', 'age': 16, 'sports': ['football', 'cricket', 'basketball']}
my_dict
```

```python
my_dict['student']  # using subscript notation
```

```python
my_dict['sports']
```

```python
# my_dict['experience']  # if the key doesn't exist, accessing using a subscript leads to a KeyError
```

```python
my_dict.get('experience')  # safer way to access keys as it returns None rather than throwing a KeyError
```

Dictionaries cannot be indexed by position

```python
# my_dict[0]
```

We can similarly access nested dictionaries by chaining subscripts

```python
class_10 = {'student1': {'name': 'Ashok Kumar', 'age': 16}, 'student2': {'name': 'Mohini Kalra', 'age': 14}, 'student3': {'name': 'Victor Savage', 'age': 15}}
class_10
```

```python
class_10['student1']
```

```python
class_10['student1']['age']
```

```python
class_10.get('student1').get('age')
```
