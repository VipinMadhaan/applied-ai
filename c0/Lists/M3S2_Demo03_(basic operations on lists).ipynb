We can find the length of a list using `len()`

```python
my_list = [9, 3, 2, 4, 5]
len(my_list)
```

```python
nested_list = [[1, 2], [3, 4, 5], [6, 7, 8, 9]]
len(nested_list)
```

```python
len(nested_list[2])
```

We can perform mathematical functions on lists using some in-built functions

```python
sum(my_list)
```

```python
min(my_list)
```

```python
max(my_list)
```

We can sort a list in ascending or descending order

```python
sorted(my_list)
```

```python
my_list
```

```python
sorted(my_list, reverse = True)
```

Arithmetic operators `+` and `*` can be overloaded to work with lists

```python
l1 = [1, 2, 3, 4, 5]
l2 = [6, 7, 8, 9, 10]
l1 + l2  # extending lists
```

```python
# l1 - l2
```

```python
# l1 * l2
```

```python
l1 * 3  # repeating/replicating a list
```

We can find the membership of elements in lists using membership operators

```python
1 in l1
```

```python
8 in l1
```

```python
2 not in l1
```
