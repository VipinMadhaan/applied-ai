Let's create some lists with elements

```python
list1 = [1, 2, 3, 4, 5]
print(list1)
```

```python
list2 = [1, 'a', True, 'abc', 3.14]  # lists can be heterogeneous in Python
print(list2)
```

We can create lists or typecast other sequences to lists by using `list()`

```python
list()
```

```python
# list(1)  # cannot convert non-iterables to lists
```

```python
list('Hello world!')
```

```python
list(range(1, 10))
```

```python
list(range(1, 10, 2))
```

```python
calculate = lambda x, y: (x + y, x - y, x * y, x / y)  # functions with multiple return values return a tuple, which is an iterable
list(calculate(4, 5))
```

The `.split()` method of strings can be used to create lists

```python
'Python programming is fun!'.split()
```
