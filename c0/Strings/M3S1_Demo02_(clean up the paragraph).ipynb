Given below is a block of text with a lot of problems. Can you clean this up so it looks alright?

````python
text = "   PyThon is an INTEResting Programming language. it IS widely used in different areas like Data Science, Web Development,   and AUTOMATION!!!    "
text
````

````python
print(text)
````

Here's the expected output:

```
Python is an interesting programming language. It is widely used in different areas like data science, web development, and automation!
```

Let's attempt to clean this up

````python
# Removing leading and trailing whitespace
cleaned_text = text.strip()
cleaned_text
````

````python
# Converting the entire text to lower case
cleaned_text = cleaned_text.lower()
cleaned_text
````

````python
# Splitting the text to get rid of excess whitespace, and rejoining with single spaces
cleaned_text = ' '.join(cleaned_text.split())
cleaned_text
````

````python
# Splitting the text into sentences (after every period), capitalising, and then joining back
sentences = cleaned_text.split('. ')

# The above data structure is a list; we can iterate through it
cleaned_text = ''
for sentence in sentences: cleaned_text += sentence.capitalize() + '. '

cleaned_text
````

````python
# Replacing unnecessary punctuation
cleaned_text = cleaned_text.replace('!!!. ', '!')
cleaned_text
````
