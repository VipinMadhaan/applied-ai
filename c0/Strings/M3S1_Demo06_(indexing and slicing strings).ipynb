We can **index** a string to retrieve a character stored at a particular position in the string. We use square brackets `[]` while indexing a string; this is known as **subscript** notation.

```python
greeting = 'Hello world!'
```

```python
# greeting(0)
```

```python
greeting[0]  # indexing to get a character using a subscript
```

```python
greeting[1]
```

```python
greeting[11]
```

```python
len(greeting)
```

```python
greeting[len(greeting) - 1]  # indexing begins from zero
```

```python
greeting[-1]  # the last element can be retrieved with a -1 index
```

```python
greeting[-2]
```

We can **slice** a string to retrieve a substring from a string

```python
greeting[:2]  # slicing to get a substring
```

```python
greeting[-2:]
```

```python
greeting[1:4]
```

```python
greeting[4:1]
```

```python
greeting[-3:-1]
```

We can additionally provide a **step** to our subscript

```python
greeting[0: len(greeting): 1]  # this is the default case; this code will retrieve every character from the string
```

```python
greeting[::1]
```

```python
greeting[::2]  # this code will retrieve every second character from the string
```

```python
greeting[::-1]  # this will retrieve every character from the string but in a reverse order
```

```python
greeting[1:-1:2]
```
