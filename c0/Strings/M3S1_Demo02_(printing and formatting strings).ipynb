We can format strings to include other variables using the `%` operator

```python
my_int = 12345  # represented by %d; 'd' stands for decimal (base 10 integers)
my_float = 3.1415  # represented by %f; 'f' stands for float

answer = 'The number I have chosen is %d and pi equals %f.' % (my_int, my_float)
print(answer)
```

```python
# Integers need not be printed in decimal; one can represent them in various ways such as octal (base 8) and hexadecimal (base 16)
print('decimal: %d, octal: %o, hexadecimal: %x' % (my_int, my_int, my_int))
```

```python
# Floats can be further formatted by providing the level of precision, for instance, %.2f will round to two digits after the decimal point
print('Pi equals %.2f' % my_float)
```

We can also format strings with other variables using the `.format()` method

```python
a = 3
b = 4.50005
total = 7.50005
answer = '{} + {} = {}'.format(a, b, total)
print(answer)
```

```python
answer = '{x} + {y} = {z}'.format(x = a, y = b, z = total)
print(answer)
```

```python
digits_after_decimal = 2
answer = '{x} + {y:.{n}f} = {z:.{n}f}'.format(x = a, y = b, z = total, n = digits_after_decimal)
print(answer)
```

We can also use **f-strings** to format strings with other variables

```python
answer = f'{a} + {b} = {total}'
print(answer)
```

```python
answer = f'{a} + {b:.{digits_after_decimal}f} = {total:.{digits_after_decimal}f}'
print(answer)
```
