We can allow programs to make decisions by using `if` statements. The `if` statement executes if the boolean value or expression provided evaluates as `True`.

```python
bool_val = 8 > 5
if bool_val: print('Eight is greater than five.')
```

```python
if True: print('This will print.')
if False: print('This will not print.')
```

```python
if 'a': print('Truthy value will print.')
if '': print('Falsy value will not print.')
if 0: print('Falsy value will not print.')
if -1: print('Truthy value will print.')
```

If we want to execute more than one statement when our `if` condition is met, we can use type these statements on multiple lines with indentation

```python
if True:
    print('Python')
    print('programming')
    print('is')
    print('fun!')
```

As long as the condition is met, it doesn't matter what's being printed

```python
if 8 > 5: print('Eight is less than five(?)')
```

We can include multiple conditions with logical operators as long as the expressions evaluate to a single boolean

```python
5 > 4
```

```python
5 < 8
```

```python
5 > 4 and 5 < 8
```

```python
if 5 > 4 and 5 < 8: print('5 is greater than 4 and less than 8.')
```

```python
if 4 < 5 < 8: print('5 is greater than 4 and less than 8.')
```

```python
is_sunny = True
temperature = 30
if is_sunny and temperature > 24: print('It is a warm and sunny day!')
```

For more complex decision making, we can use multiple `if` statements in a sequence

```python
marks = 80
if marks >= 90: print('Grade A')
if marks >= 80: print('Grade B')
if marks >= 70: print('Grade C')
if marks >= 60: print('Grade D')
if marks < 60: print('Fail')
```

That doesn't look right! Remember, your code cannot think for itself and you need to be very explicit in telling it what you want it to do.

```python
if marks >= 90: print('Grade A')
if marks >= 80 and marks < 90: print('Grade B')
if marks >= 70 and marks < 80: print('Grade C')
if marks >= 60 and marks < 70: print('Grade D')
if marks < 60: print('Fail')
```
