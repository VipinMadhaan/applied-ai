Higher-order functions are functions which accept other functions as arguments

```python
# Defining some functions to perform operations on two variables
def add(x, y): return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
```

```python
def calculate(operation, x, y):  # this is a higher-order function which accepts another function as an argument to its operation parameter
    return operation(x, y)
```

```python
print(calculate(add, 4, 4))
```

```python
print(calculate(subtract, 5, 7))
```

```python
print(calculate(multiply, 3, 2))
```

```python
print(calculate(lambda x, y: x / y, 20, 5))
```
