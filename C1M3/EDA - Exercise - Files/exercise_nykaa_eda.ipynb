```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Task 1
## Inspect the data
- Load the data
- Study the shape of the data
- Check the data for data types and missing values
- Display top 5 rows
- Study the summary statistics of the data using `describe()` and include all data types in the result

```python
df = pd.read_csv('/content/nykaa_eda.csv')

print('Shape of the data:')
display(df.shape)

print('\nData types and missing values:')
display(df.info())

print('\nTop 5 rows of the data:')
display(df.head())

print('\nSummary statistics of the data (including all data types):')
display(df.describe(include='all'))
```

# Task 2
## Analyse product price distribution
- Plot the discrete histogram of `'product_price'` with 5 bins
- Look for trends in the distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['product_price'], bins=5, kde=False)
plt.title('Distribution of Product Price (5 Bins)')
plt.xlabel('Product Price')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

# Task 3
## Analyse product brand counts
- Obtain the counts of all the brands in the dataset
- Find the top 5 brands based on the count in descending order
- Create a copy of the data frame with only these identified brands
- Check the shape of the data subset that you created

```python
print('Counts of all brands:')
brand_counts = df['product_brand'].value_counts()
display(brand_counts)

print('\nTop 5 brands:')
top_5_brands = brand_counts.head(5).index.tolist()
display(top_5_brands)

print('\nCreating a subset dataframe with only the top 5 brands.')
df_subset = df[df['product_brand'].isin(top_5_brands)].copy()

print('\nShape of the new subset dataframe:')
display(df_subset.shape)
```

**Note:** From this point onwards, please use the subset data that you created in the previous step

# Task 4
## Analyse price distribution by brand
- Obtain the mean product price of each brand

```python
print('Mean product price for each brand (using df_subset):')
mean_price_by_brand = df_subset.groupby('product_brand')['product_price'].mean()
display(mean_price_by_brand)
```

# Task 5
## Analyse price distribution by product category
- Obtain the mean product price of each product category
- Sort the result in ascending order

```python
print('Mean product price for each product category (using df_subset), sorted in ascending order:')
mean_price_by_category = df_subset.groupby('product_category')['product_price'].mean().sort_values(ascending=True)
display(mean_price_by_category)
```

# Task 6
## Analyse product ratings by presence of reviews

```python
print('Mean product ratings by presence of reviews (using df_subset):')
mean_rating_by_reviews = df_subset.groupby('has_reviews')['product_rating'].mean()
display(mean_rating_by_reviews)
```

# Task 8
## Analyse relation between review counts and ratings
- Plot a scatter plot between `'product_reviews_count'` and `'product_rating'`
- Plot another scatter plot between natural logarithm of `'product_reviews_count'` and `'product_rating'`

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x='product_reviews_count', y='product_rating', data=df_subset)
plt.title('Product Reviews Count vs. Product Rating')
plt.xlabel('Product Reviews Count')
plt.ylabel('Product Rating')

plt.subplot(1, 2, 2)
sns.scatterplot(x=np.log1p(df_subset['product_reviews_count']), y='product_rating', data=df_subset)
plt.title('Natural Log of Product Reviews Count vs. Product Rating')
plt.xlabel('Natural Log of Product Reviews Count')
plt.ylabel('Product Rating')

plt.tight_layout()
plt.show()
```

# Task 9
## Analyse relation between product ratings and price
- Plot a scatter plot between `'product_rating'` and `'product_price'`
- Plot another scatter plot between `'product_rating'` and natural logarithm of `'product_price'`
- Plot a regression plot using `regplot` from `seaborn` between `'product_rating'` and natural logarithm of `'product_price'`

```python
plt.figure(figsize=(18, 6))

# Plot 1: Scatter plot between 'product_rating' and 'product_price'
plt.subplot(1, 3, 1)
sns.scatterplot(x='product_rating', y='product_price', data=df_subset)
plt.title('Product Rating vs. Product Price')
plt.xlabel('Product Rating')
plt.ylabel('Product Price')

# Plot 2: Scatter plot between 'product_rating' and natural logarithm of 'product_price'
plt.subplot(1, 3, 2)
sns.scatterplot(x='product_rating', y=np.log1p(df_subset['product_price']), data=df_subset)
plt.title('Product Rating vs. Natural Log of Product Price')
plt.xlabel('Product Rating')
plt.ylabel('Natural Log of Product Price')

# Plot 3: Regression plot between 'product_rating' and natural logarithm of 'product_price'
plt.subplot(1, 3, 3)
sns.regplot(x='product_rating', y=np.log1p(df_subset['product_price']), data=df_subset)
plt.title('Regression Plot: Rating vs. Natural Log of Price')
plt.xlabel('Product Rating')
plt.ylabel('Natural Log of Product Price')

plt.tight_layout()
plt.show()
```

# Task 10
## Analyse relation between review counts, ratings, and brand
- Plot a scatter plot between natural logarithm of `'product_reviews_count'` and `'product_rating'` and set hue as `'product_brand'`
- Plot a regression plot using `regplot` from `seaborn` between natural logarithm of `'product_reviews_count'` and `'product_rating'` only for the brand `'LIME CRIME'`
- Plot another regression plot using `regplot` from `seaborn` between natural logarithm of `'product_reviews_count'` and `'product_rating'` only for the brand `'LAKME'`

```python
plt.figure(figsize=(18, 6))

# Plot 1: Scatter plot between natural logarithm of 'product_reviews_count' and 'product_rating' with hue as 'product_brand'
plt.subplot(1, 3, 1)
sns.scatterplot(x=np.log1p(df_subset['product_reviews_count']), y='product_rating', hue='product_brand', data=df_subset)
plt.title('Natural Log of Reviews Count vs. Product Rating by Brand')
plt.xlabel('Natural Log of Product Reviews Count')
plt.ylabel('Product Rating')

# Plot 2: Regression plot for 'LIME CRIME'
plt.subplot(1, 3, 2)
sns.regplot(x=np.log1p(df_subset[df_subset['product_brand'] == 'LIME CRIME']['product_reviews_count']), y=df_subset[df_subset['product_brand'] == 'LIME CRIME']['product_rating'], scatter_kws={'alpha':0.6})
plt.title('Regression Plot: LIME CRIME (Log Reviews Count vs. Rating)')
plt.xlabel('Natural Log of Product Reviews Count')
plt.ylabel('Product Rating')

# Plot 3: Regression plot for 'LAKME'
plt.subplot(1, 3, 3)
sns.regplot(x=np.log1p(df_subset[df_subset['product_brand'] == 'LAKME']['product_reviews_count']), y=df_subset[df_subset['product_brand'] == 'LAKME']['product_rating'], scatter_kws={'alpha':0.6})
plt.title('Regression Plot: LAKME (Log Reviews Count vs. Rating)')
plt.xlabel('Natural Log of Product Reviews Count')
plt.ylabel('Product Rating')

plt.tight_layout()
plt.show()
```

# Task 11
## Analyse relation between ratings, price, and brand
- Plot a scatter plot between `'product_rating'` and `'product_price'` and set hue as `'product_brand'`
- Plot a scatter plot between `'product_rating'` and natural logarithm of `'product_price'` and set hue as `'product_brand'`
- Plot a regression plot using `regplot` from `seaborn` between `'product_rating'` and natural logarithm of `'product_price'` only for the brand `'HIMALAYA'`
- Plot a regression plot using `regplot` from `seaborn` between `'product_rating'` and natural logarithm of `'product_price'` only for the brand `'LIME CRIME'`

```python
plt.figure(figsize=(24, 6))

# Plot 1: Scatter plot between 'product_rating' and 'product_price' with hue as 'product_brand'
plt.subplot(1, 4, 1)
sns.scatterplot(x='product_rating', y='product_price', hue='product_brand', data=df_subset)
plt.title('Product Rating vs. Product Price by Brand')
plt.xlabel('Product Rating')
plt.ylabel('Product Price')

# Plot 2: Scatter plot between 'product_rating' and natural logarithm of 'product_price' with hue as 'product_brand'
plt.subplot(1, 4, 2)
sns.scatterplot(x='product_rating', y=np.log1p(df_subset['product_price']), hue='product_brand', data=df_subset)
plt.title('Product Rating vs. Natural Log of Price by Brand')
plt.xlabel('Product Rating')
plt.ylabel('Natural Log of Product Price')

# Plot 3: Regression plot for 'HIMALAYA'
himalaya_data = df_subset[df_subset['product_brand'] == 'HIMALAYA']
plt.subplot(1, 4, 3)
sns.regplot(x=himalaya_data['product_rating'], y=np.log1p(himalaya_data['product_price']), scatter_kws={'alpha':0.6})
plt.title('Regression Plot: HIMALAYA (Rating vs. Log Price)')
plt.xlabel('Product Rating')
plt.ylabel('Natural Log of Product Price')

# Plot 4: Regression plot for 'LIME CRIME'
lime_crime_data = df_subset[df_subset['product_brand'] == 'LIME CRIME']
plt.subplot(1, 4, 4)
sns.regplot(x=lime_crime_data['product_rating'], y=np.log1p(lime_crime_data['product_price']), scatter_kws={'alpha':0.6})
plt.title('Regression Plot: LIME CRIME (Rating vs. Log Price)')
plt.xlabel('Product Rating')
plt.ylabel('Natural Log of Product Price')

plt.tight_layout()
plt.show()
```
