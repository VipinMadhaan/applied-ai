# Task 1
## Inspect the data
- Load the data
- Display the first 5 rows
- Display 5 random rows
- Check the number of rows and columns
- Inspect the data types of all features

```python
df = pd.read_csv('/content/nykaa_data_preprocessing.csv')

print('First 5 rows:')
display(df.head())

print('\n5 random rows:')
display(df.sample(5))

print('\nNumber of rows and columns:')
print(df.shape)

print('\nData types of all features:')
display(df.info())
```

# Task 2
## Fix data types
- Ensure that all features are stored in the most suitable data type
- Convert the `'Product Price'` feature to `float64` data type

```python
# Identify columns with non-suitable data types and convert them

# Convert 'Product Price' to float64
# First, clean the 'Product Price' column by removing non-numeric characters
# Assuming currency symbols like '₹' and commas ',' might be present
df['Product Price'] = df['Product Price'].astype(str).str.replace('₹', '', regex=False)
df['Product Price'] = df['Product Price'].astype(str).str.replace(',', '', regex=False)

# Convert to float64, coercing errors to NaN
df['Product Price'] = pd.to_numeric(df['Product Price'], errors='coerce')

# Display data types again to confirm changes
print('Updated data types after type conversion:')
display(df.info())
```

# Task 3
## Handle missing values
- Identify columns with missing values
- Replace missing values in string columns with `'nodata'`
- Replace missing numeric values with median

```python
# Identify columns with missing values
print('Missing values before handling:')
print(df.isnull().sum())

# Separate columns into numeric and string/object types
numeric_cols = df.select_dtypes(include=['number']).columns
string_cols = df.select_dtypes(include=['object']).columns

# Replace missing values in string columns with 'nodata'
for col in string_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna('nodata')

# Replace missing numeric values with the median
for col in numeric_cols:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)

# Display missing values after handling to confirm
print('\nMissing values after handling:')
print(df.isnull().sum())
```

# Task 4
## Remove duplicate records
- Check whether duplicate rows exist
- Remove duplicate records
- Display how many rows were removed

```python
# Check for duplicate rows
duplicate_rows_before = df.duplicated().sum()
print(f"Number of duplicate rows before removal: {duplicate_rows_before}")

# Remove duplicate records
df.drop_duplicates(inplace=True)

# Display how many rows were removed
duplicate_rows_after = df.duplicated().sum()
removed_rows = duplicate_rows_before - duplicate_rows_after
print(f"Number of duplicate rows after removal: {duplicate_rows_after}")
print(f"Number of rows removed: {removed_rows}")
```

# Task 5
## Standardise column names
- Remove any leading and trailing whitespaces from column names
- Convert all column names to lowercase
- Replace spaces between words with underscores (`_`)
- Ensure there are no multiple consecutive underscores
- Remove any leading or trailing underscores
- Display the updated column names

```python
# Standardise column names

# Get current column names
current_columns = df.columns

# Create a list to hold new column names
new_columns = []

for col in current_columns:
    # 1. Remove any leading and trailing whitespaces
    col = col.strip()
    # 2. Convert all column names to lowercase
    col = col.lower()
    # 3. Replace spaces between words with underscores
    col = col.replace(' ', '_')
    # 4. Ensure there are no multiple consecutive underscores
    col = '_'.join(filter(None, col.split('_')))
    # 5. Remove any leading or trailing underscores (after joining, filter(None) already handles multiple underscores reducing to single and removing leading/trailing if they result from empty strings after split)
    col = col.strip('_')
    new_columns.append(col)

# Assign the new column names back to the DataFrame
df.columns = new_columns

# Display the updated column names
print('Updated column names:')
print(df.columns)
```

# Task 6
## Engineer new features
- Create a column `'price_category'`:
    - `'low'` if product price is less than 500
    - `'medium'` if product price is between 500 and 2000 (both inclusive)
    - `'high'` if product price is greater than 2000
- Create a column `'has_reviews'`:
    - `'yes'` if the product has at least one review
    - `'no'` otherwise
- Obtain the mode of the `'price_category'` feature

```python
# Create 'price_category' column
def get_price_category(price):
    if price < 500:
        return 'low'
    elif 500 <= price <= 2000:
        return 'medium'
    else:
        return 'high'

df['price_category'] = df['product_price'].apply(get_price_category)

# Create 'has_reviews' column
df['has_reviews'] = df['product_reviews_count'].apply(lambda x: 'yes' if x > 0 else 'no')

# Obtain the mode of the 'price_category' feature
price_category_mode = df['price_category'].mode()[0]

print(f"Mode of 'price_category': {price_category_mode}")
print("\nFirst 5 rows with new features:")
display(df[['product_price', 'price_category', 'product_reviews_count', 'has_reviews']].head())
```

# Task 7
## Treat outliers
- Use the interquartile range (IQR) method to identify outliers in the `'product_price'` column
  - IQR = Q3 - Q1, UL = Q3 + 1.5 * IQR, LL = Q1 - 1.5 * IQR
- Find the number of outliers according to the IQR method and treat them by capping them to the UL or the LL accordingly

```python
# Treat outliers in 'product_price' using IQR method

# Calculate Q1, Q3, and IQR for 'product_price'
Q1 = df['product_price'].quantile(0.25)
Q3 = df['product_price'].quantile(0.75)
IQR = Q3 - Q1

# Calculate Upper Limit (UL) and Lower Limit (LL)
UL = Q3 + 1.5 * IQR
LL = Q1 - 1.5 * IQR

print(f"Q1 for product_price: {Q1}")
print(f"Q3 for product_price: {Q3}")
print(f"IQR for product_price: {IQR}")
print(f"Upper Limit (UL) for product_price: {UL}")
print(f"Lower Limit (LL) for product_price: {LL}")

# Find the number of outliers
outliers = df[(df['product_price'] < LL) | (df['product_price'] > UL)]
num_outliers = len(outliers)
print(f"\nNumber of outliers in 'product_price': {num_outliers}")

# Treat outliers by capping them to UL or LL
df['product_price'] = df['product_price'].clip(lower=LL, upper=UL)

# Verify that outliers have been treated (optional, but good for checking)
num_outliers_after = len(df[(df['product_price'] < LL) | (df['product_price'] > UL)])
print(f"Number of outliers in 'product_price' after capping: {num_outliers_after}")

print("\nFirst 5 rows of 'product_price' after outlier treatment:")
display(df[['product_price']].head())
```
