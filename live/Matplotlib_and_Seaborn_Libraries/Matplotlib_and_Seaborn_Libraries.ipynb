## Why use Python visualization libraries when Power BI and Tableau already exist?

**Power BI and Tableau are great for dashboards and business reporting.**  
But **Matplotlib and Seaborn are useful when you want more control, flexibility, and integration with data analysis in Python.**

---

### Key reasons:

**1. They work directly inside Python workflows**  
If you are already cleaning, transforming, and analyzing data in Python using Pandas, it is easier to visualize in the same environment.

**2. Better for analysts and data scientists**  
In real projects, people often need to:

- clean data  
- apply logic  
- build models  
- visualize results  

All in one notebook or script.

**3. Highly customizable**  
You can control titles, labels, colors, layouts, annotations, subplots, axes, and more.

**4. Useful for exploratory data analysis**  
Before building dashboards, analysts often use Matplotlib and Seaborn to quickly understand the data.

**5. Essential in machine learning and data science workflows**  
For example:

- distribution of target variable  
- feature relationships  
- model error visualization  
- confusion matrix  
- residual plots  

---

## Simple comparison table

| Tool | Best For | Limitation |
|------|---------|-----------|
| Excel | Quick manual charts | Limited for larger workflows |
| Power BI / Tableau | Dashboards and business reporting | Less flexible for code-driven analysis |
| Matplotlib / Seaborn | Analysis, experimentation, custom visualization in Python | Requires coding |

---

## Introducing Matplotlib and Seaborn

---

### What is Matplotlib?

Matplotlib is the foundational plotting library in Python.

> **“If Python visualization had a base engine, Matplotlib would be that engine.”**

It is used for:

- line charts  
- bar charts  
- scatter plots  
- histograms  
- pie charts  
- full customization  

---

### What is Seaborn?

Seaborn is built on top of Matplotlib.

> **“Seaborn makes statistical visualization easier and prettier.”**

It is useful for:

- attractive charts with less code  
- better default styling  
- statistical relationships  
- grouped comparisons  
- heatmaps  
- boxplots  

---

### Difference between Matplotlib and Seaborn

| Matplotlib | Seaborn |
|------------|--------|
| More control | Easier and cleaner for many common charts |
| Slightly more code | Less code |
| Basic foundation library | Built on top of Matplotlib |
| Good for custom designs | Good for attractive statistical charts |

# **Line Chart**

### When to use

Use a **line chart** when you want to show **change over time**.

---

### Industry examples

- **Retail:** Monthly sales trend across stores  
- **Finance:** Daily stock price movement  
- **Healthcare:** Number of patient visits over months  
- **EdTech:** Weekly active learners on a platform  
- **Manufacturing:** Machine downtime trend by week  

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [120, 150, 170, 160, 180, 210],
    'Profit': [20, 25, 30, 28, 35, 40],
    'Category': ['A', 'B', 'A', 'B', 'A', 'B']
}

df = pd.DataFrame(data)
print(df)

import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 150, 170, 160, 180, 210]

plt.plot(months, sales, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()

# x-axis usually shows time
# y-axis shows the changing metric
# helps identify growth, decline, or fluctuation
```

```python
#Creating a new column for year
transactions['Order Year'] = pd.DatetimeIndex(transactions['Order Date']).year
transactions[['Order Date', 'Order Year', 'Sales']]
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
df = pd.DataFrame(transactions.groupby('Order Year')['Sales'].sum())
df
```

```python
#plt.ticklabel_format(style='plain', axis='y')
plt.plot(df.index, df['Sales'], 'g*-')
plt.xticks(df.index, df.index.astype(int))
plt.xlabel('Order Year', size = 15)
plt.ylabel('Sales ($)', size = 15)
plt.title('Yearwise Sales Trend', size = 20)
```

### Interpretation

- Sales remained relatively stable between 2020 and 2021, indicating a period of stagnation  
- A noticeable increase is observed from 2021 to 2022, suggesting recovery or growth  
- There is a sharp rise in 2023, indicating strong business performance or expansion  
- Overall, the trend shows a **positive upward trajectory**, especially in the later years  
- The significant jump in 2023 may warrant further analysis to understand key growth drivers  

```python
#Creating a new column for month
transactions['Order Month'] = pd.DatetimeIndex(transactions['Order Date']).month
transactions[['Order Date', 'Order Month', 'Profit']]

pd.set_option('display.float_format', '{:.2f}'.format)
df = pd.DataFrame(transactions.groupby('Order Month')['Profit'].sum())

plt.ticklabel_format(style='plain', axis='y')
plt.plot(df.index, df['Profit'], 'rX-')
plt.xlabel('Month', size = 15)
plt.ylabel('Profit', size = 15)
plt.title('Monthwise Profit Trend', size = 20)
```

```python
sns.lineplot(x='Order Month', y='Profit', data=df, marker='o', color = 'red')
```

### Interpretation

- Profit shows significant fluctuation across months, indicating inconsistent performance  
- There is a sharp dip into negative profit around mid-year, suggesting a potential loss period  
- A strong recovery is observed after this dip, with profits rising sharply in subsequent months  
- The highest profit is recorded around Month 9, indicating peak business performance  
- Overall, the trend highlights **high volatility**, which may require investigation into cost, demand, or seasonal factors  

# **Bar Chart**

### When to use

Use a **bar chart** when you want to **compare categories**.

---

### Industry examples

- **E-commerce:** Sales by product category  
- **HR:** Number of employees by department  
- **Healthcare:** Patients by disease type  
- **Banking:** Loan applications by city  
- **Education:** Student enrollments by course  

```python
categories = ['Electronics', 'Clothing', 'Groceries', 'Furniture']
values = [500, 300, 450, 200]

plt.bar(categories, values)
plt.title('Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()

# each bar is a category
# bar height shows value
# best when comparing discrete groups
```

### Interpretation

- Electronics has the highest sales, indicating it is the top-performing category  
- Groceries also shows strong performance, contributing significantly to overall sales  
- Clothing has moderate sales, suggesting steady but not leading performance  
- Furniture has the lowest sales among all categories, indicating potential underperformance  
- Overall, sales are unevenly distributed across categories, highlighting opportunities to improve weaker segments  

```python
df = pd.DataFrame(transactions.groupby('Sub-Category')['Sales'].mean())
print(df)
bars = plt.bar(df.index, df['Sales'])
plt.xlabel('Sub-Category', size = 15)
plt.ylabel('Average Sales', size = 15)
plt.title('Sub-Category wise Average Sales', size = 20)
plt.xticks(rotation = 90)
plt.bar_label(bars, fmt='%.2f')
```

### Interpretation

- Copiers have the highest average sales by a significant margin, making it the top-performing sub-category  
- Machines also show strong performance, contributing substantially to overall sales  
- Tables, Chairs, and Bookcases fall in the mid-range, indicating stable demand  
- Several sub-categories like Labels, Fasteners, and Art have very low average sales, suggesting limited contribution  
- The distribution is highly skewed, with a few sub-categories driving most of the sales  
- This indicates a need to focus on high-performing sub-categories while exploring strategies to improve underperforming ones  

```python
df = transactions.groupby(['Category', 'Sub-Category'], as_index=False)['Sales'].mean()
df = df.sort_values(by=['Category', 'Sales'], ascending=[True, False])
df
```

```python
category_colors = {
    'Furniture': 'brown',
    'Technology': 'lightblue',
    'Office Supplies': 'purple'
}

category_order = df.groupby('Category')['Sales'].mean().sort_values(ascending=False).index

sns.barplot(
    x='Sub-Category',
    y='Sales',
    hue='Category',              # 👈 IMPORTANT CHANGE
    data=df,
    palette=category_colors      # 👈 map directly
)

plt.xticks(rotation=45, ha='right')
plt.title("Sales by Category and Sub-category")
plt.show()
```

### Interpretation

- Technology category significantly outperforms others, with Copiers and Machines showing the highest sales  
- Furniture category shows moderate performance, with Tables and Chairs leading within the category  
- Office Supplies has the lowest overall sales contribution, with most sub-categories performing at relatively lower levels  
- There is a clear imbalance in sales distribution, with Technology driving a major share of total sales  
- A few sub-categories dominate within each category, while several others contribute minimally  
- This suggests an opportunity to strengthen underperforming sub-categories, especially within Office Supplies  

# **Histogram**

### When to use

Use a **histogram** to understand the **distribution of a numeric variable**.

---

### Industry examples

- **Banking:** Distribution of customer ages  
- **Insurance:** Distribution of claim amounts  
- **EdTech:** Distribution of student marks  
- **Retail:** Distribution of order values  
- **Manufacturing:** Distribution of product weights  



```python
ages = [22, 25, 27, 28, 30, 30, 31, 35, 36, 40, 42, 45, 48, 50, 52]

plt.hist(ages, bins=5)
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# histogram is not for categories
# it groups numeric values into ranges
# useful for understanding concentration, spread, skewness
```

### Interpretation

- The age distribution appears fairly spread out across different age groups, indicating a diverse customer base  
- A higher concentration of customers is observed in the age range of approximately 25 to 35  
- There is a gradual presence of customers in the 40+ age group, though slightly less frequent  
- No extreme skewness is observed, suggesting a relatively balanced distribution  
- There are no clear outliers, indicating that most customers fall within a reasonable age range  

```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)
plt.hist(transactions['Sales'], bins= np.arange(0,23500,100))
plt.xlabel('Histogram Bins for Sales', size= 15)
plt.ylabel('Frequency Counts', size= 15)
```

### Interpretation

- The distribution of sales is highly skewed towards lower values, with a large concentration near the lower range  
- A small number of transactions have significantly higher sales, creating a long right tail (positive skewness)  
- Most sales transactions are of low value, while high-value transactions are relatively rare  
- The presence of extreme high values suggests potential outliers or premium transactions  
- This indicates that the business may rely on a high volume of smaller transactions, with occasional large sales contributing to overall revenue  


# **Scatter Plot**

### When to use

Use a **scatter plot** to understand the **relationship between two numeric variables**.

---

### Industry examples

- **Marketing:** Ad spend vs sales generated  
- **Real Estate:** House size vs price  
- **Healthcare:** BMI vs blood pressure  
- **Education:** Study hours vs test scores  
- **Manufacturing:** Machine temperature vs defect rate  


```python
ad_spend = [10, 20, 30, 40, 50, 60]
sales = [100, 120, 140, 160, 180, 210]

plt.scatter(ad_spend, sales)
plt.title('Ad Spend vs Sales')
plt.xlabel('Ad Spend')
plt.ylabel('Sales')
plt.show()

# each dot is one observation
# helps identify positive, negative, or no relationship
# good for spotting clusters and outliers
```

```python
plt.scatter(transactions['Sales'], transactions['Profit'], marker='^', c='green')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.title('Scatter Plot')
```

### Interpretation

- There is a general positive relationship between sales and profit, indicating that higher sales tend to result in higher profits  
- However, the relationship is not perfectly linear, as there is significant variability in profit at similar sales levels  
- Several points show negative profit despite moderate sales, suggesting potential losses or high costs in certain transactions  
- A few high-sales transactions generate very high profits, indicating strong performance in specific cases  
- The presence of outliers (both high profit and high loss) suggests variability in pricing, discounts, or cost structures  

# **Box Plot**

### When to use

Use a **box plot** to understand **spread, median, and outliers**.

---

### Industry examples  

- **Retail:** Order value distribution across regions
- **HR:** Salary distribution by department
- **Healthcare:** Recovery time across treatment groups
- **Education:** Marks distribution across classes
- **Logistics:** Delivery time across cities

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df_box = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'Sales', 'HR', 'HR', 'HR', 'Tech', 'Tech', 'Tech'],
    'Salary': [40000, 45000, 70000, 35000, 37000, 39000, 60000, 65000, 120000]
})

sns.boxplot(x='Department', y='Salary', data=df_box)
plt.title('Salary Distribution by Department')
plt.show()

# line inside box = median
# box = middle 50 percent of data
# whiskers = spread
# dots outside = possible outliers
```

### Interpretation

- The Tech department has the highest salary levels, with a wider spread indicating significant variation in pay  
- Sales shows moderate salary levels with some variability, suggesting differences based on experience or performance  
- HR has the lowest and most tightly clustered salary range, indicating relatively uniform compensation  
- The presence of a high upper range in Tech suggests senior roles or specialized skill premiums  
- Overall, salary distribution varies significantly across departments, with Tech being the most diverse and highest paying  

```python
regions = transactions['Region'].unique()
data = [transactions[transactions['Region'] == region]['Sales'] for region in regions]

# Plot
plt.boxplot(data, labels=regions)
plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

### Interpretation

- Sales are highly concentrated at lower values across all regions, with a large number of small transactions  
- All regions show significant outliers, indicating occasional high-value sales  
- The South region has the highest extreme value, suggesting presence of very large transactions  
- Central and West also show notable high-value outliers, while East appears relatively moderate  
- The overall distribution is highly skewed, indicating that a few large transactions contribute disproportionately to total sales

# **Heatmap**

### When to use

Use a **heatmap** to show **intensity or relationship in matrix form**.

---

### Industry examples  

- **Retail:** Sales by city and product category
- **Education:** Student performance by subject
- **HR:** Attrition rates across teams and locations
- **Banking:** Correlation between financial variables
- **Healthcare:** Hospital occupancy by ward and month

```python
import numpy as np

data = np.array([
    [50, 60, 70],
    [80, 55, 65],
    [90, 75, 85]
])

sns.heatmap(data, annot=True, cmap='YlGnBu',
            xticklabels=['Product A', 'Product B', 'Product C'],
            yticklabels=['North', 'South', 'West'])

plt.title('Sales Heatmap by Region and Product')
plt.show()

# darker or brighter areas show stronger values
# useful when there are many combinations to compare
# easier than reading long tables
```

### Interpretation

- The West region shows the highest sales across all products, indicating strong overall performance  
- Product A performs exceptionally well in the West and South regions, making it a key revenue driver  
- Product B shows relatively lower sales across regions, suggesting weaker demand compared to other products  
- Product C performs consistently across regions, with strong contribution especially in the West  
- The North region has comparatively lower sales across all products, indicating potential growth opportunities  

# **Pie Chart**

### When to use

Use a **pie chart** to show parts of a whole.
Use it only when there are **few categories.** Too many slices make it confusing.

---

### Industry examples  

- **Marketing:** Lead source contribution
- **Retail:** Revenue share by product category
- **Banking:** Loan portfolio split by segment
- **Education:** Enrollment share by program
- **Healthcare:** Patient share by treatment type

```python
labels = ['Online', 'Referral', 'Social Media', 'Email']
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Lead Source Contribution')
plt.show()

# best for percentage contribution
# not good when too many categories exist
# bar chart is often clearer than pie chart
```

### Interpretation

- Online is the largest contributor, accounting for 40% of total leads, making it the most effective channel  
- Referral contributes 25%, indicating strong performance driven by existing customer networks  
- Social Media contributes 20%, showing moderate effectiveness in lead generation  
- Email has the lowest contribution at 15%, suggesting relatively lower impact compared to other channels  
- Overall, lead generation is dominated by digital channels, with Online and Social Media playing a key role  

```python
orders.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.2f%%',
                                             colors= ['red', 'green', 'purple', 'yellow'])
```

### Interpretation

- West contributes the highest share of total sales, accounting for approximately 31.8%, making it the top-performing region  
- East follows closely with around 29.7%, indicating strong performance  
- Central contributes about 21.6%, showing moderate sales contribution  
- South has the lowest share at approximately 16.8%, indicating comparatively weaker performance  
- Overall, sales distribution is uneven across regions, with West and East driving the majority of revenue  

```python
labels = 'West' , 'East', 'South', 'Central'
order_counts = orders.Region.value_counts()
colors = ['pink', 'yellow', 'green', 'blue']
explode = (0.2, 0.1, 0.1, 0.1)
plt.pie(order_counts, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.title("Percentage of orders placed in each region",fontsize=20)
plt.legend( labels, loc="best")
plt.plot()
```

### Interpretation

- West has the highest share of orders at approximately 31.9%, making it the leading region in terms of order volume  
- East follows closely with around 29.3%, indicating strong demand  
- South contributes about 22.9%, showing moderate order activity  
- Central has the lowest share at approximately 15.9%, indicating comparatively lower order volume  
- Overall, order distribution is uneven across regions, with West and East contributing the majority of orders  

# Simple decision guide

|Question|Best Graph|
|--------|----------|
|Do I want to show trend over time?|Line chart|
|Do I want to compare categories?|Bar chart|
|Do I want to see distribution of one numeric variable?|Histogram|
|Do I want to see relation between two numeric variables?|Scatter plot|
|Do I want to spot spread and outliers?|Box plot|
|Do I want to compare values in a matrix?|Heatmap|
|Do I want to show percentage share?|Pie chart|


**The right graph depends on the question you are asking from the data.**
