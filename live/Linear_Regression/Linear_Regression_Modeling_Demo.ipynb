# <center>Linear Regression

Linear regression is a supervised learning algorithm that models the relationship between one or more independent variables (predictors) and a dependent variable (outcome). It assumes that this relationship is linear, meaning the change in the dependent variable is proportional to the change in the independent variables.

![](https://www.biomedware.com/files/documentation/spacestat/Statistics/Multivariate_Modeling/Regression/regression_line.png)

### Simple Linear Regression:
Simple linear regression is used when there is one independent variable $x$ and one dependent variable $y$. The goal is to fit a straight line that best represents the relationship between these two variables.

The mathematical form of simple linear regression is:
$$y = \beta_0 + \beta_1 x + \epsilon$$
Where:
- $y$ is the dependent variable (outcome),
- $x$ is the independent variable (predictor),
- $\beta_0$ is the **intercept** (the value of $y$ when $x = 0$),
- $\beta_1$ is the **slope** (how much $y$ changes for a one-unit change in $x$),
- $\epsilon$ is the **error term** (the difference between actual and predicted $y$).



### Example: Predicting Building Height from the Number of Stories

Let's consider the relationship between the number of stories a building has and its height. For this example, we assume that every story adds a constant amount to the building's height. In reality, this could be approximately true if each story is about the same height (say, around 3 meters per story).

#### Step 1: Problem Setup
We want to predict the height of a building based on the number of stories it has. In this case:
- **Independent variable (X)**: The number of stories in the building.
- **Dependent variable (Y)**: The height of the building in meters.

### Step 2: Assumptions
In linear regression, we assume that the relationship between the number of stories and the building's height is linear. This means that for each additional story, the height increases by a constant amount. For example:
- A building with 5 stories might be around 15 meters tall.
- A 10-story building might be around 30 meters tall.
- A 20-story building might be around 60 meters tall.

The constant increase per story represents the slope of the linear relationship. Here, we assume that each story adds approximately 3 meters to the building's total height.

### Step 3: Equation of the Line
In linear regression, we model the relationship between the number of stories and the height using the following equation:
$$Y = \beta_0 + \beta_1 X$$
Where:
- $Y$ is the height of the building (in meters).
- $X$ is the number of stories.
- $\beta_0$ is the intercept (the predicted height of a building when it has 0 stories, which might be 0 or a small value representing base construction height).
- $\beta_1$ is the slope (the increase in height for each additional story).

### Step 4: Fitting the Model
Using historical data of different buildings with known heights and number of stories, we can fit a simple linear regression model. The model will estimate the parameters:
- $\beta_0$, the intercept.
- $\beta_1$, the slope (in this case, the height added per story).

For example, if we observe that a building with 5 stories is 15 meters tall, and a 10-story building is 30 meters tall, the model will find that each additional story adds roughly 3 meters to the height.

### Step 5: Interpretation
Once the model is trained, we can interpret the results:
- **Intercept ($\beta_0$)**: If the intercept is 0, this means that a building with 0 stories has 0 height (which makes sense in this case). If it’s a small positive value, it may represent the base construction height before stories are added.
- **Slope ($\beta_1$)**: The slope would tell us the height added per story. If $\beta_1 = 3$, this means each additional story adds 3 meters to the height.

### Step 6: Using the Model for Prediction
After the model is built, we can use it to predict the height of buildings with any number of stories. For instance:
- A 25-story building would be predicted to have a height of approximately $25 \times 3 = 75$ meters.
- A 40-story building would be predicted to have a height of approximately $40 \times 3 = 120$ meters.

### Multiple Linear Regression:
In multiple linear regression, more than one independent variable is used to predict the dependent variable. For example, predicting sales based on TV, radio, and newspaper advertising expenditures:
$$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n + \epsilon$$
Where:
- $x_1, x_2, \ldots, x_n$ are the independent variables,
- $\beta_0, \beta_1, \ldots, \beta_n$ are the coefficients.

# How Does Linear Regression Work?

The key idea behind linear regression is to find the line (or hyperplane) that best fits the data points. This is done using the **Ordinary Least Squares (OLS)** method, OLS is a classical method to estimate coefficients in linear regression, but there are other methods like Gradient Descent, MLE, and regularized methods such as Ridge and Lasso.

![](https://www.biomedware.com/files/documentation/spacestat/Statistics/Multivariate_Modeling/Regression/regression_line.png)

## Ordinary Least Squares (OLS) Method:
OLS estimates the coefficients of the regression model ($\beta_0, \beta_1, \ldots$) by minimizing the sum of the squared residuals (errors). A residual is the difference between the actual values $y_i$ and the predicted values $\hat{y_i}$:
$$\text{Residual}_i = y_i - \hat{y}_i$$
The goal of OLS is to minimize the **sum of squared residuals (SSR)**:
$$SSR = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

## Gradient Descent

Gradient Descent is an optimization algorithm used for finding the minimum of a function iteratively. In the context of linear regression, Gradient Descent can be used to minimize the cost function (e.g., the sum of squared residuals) without solving the system of equations directly.

Gradient Descent is particularly useful for large datasets or when the normal equations in OLS become computationally expensive to solve due to matrix inversions. It works by updating the coefficients iteratively:
$$\beta_j := \beta_j - \alpha \frac{\partial J(\beta)}{\partial \beta_j}$$
Where:
- $\alpha$ is the learning rate.
- $J(\beta)$ is the cost function (typically the sum of squared errors in linear regression).

**Stochastic Gradient Descent (SGD)** is a variation where the updates are made for individual training examples rather than the whole dataset.

## Assumptions of Linear Regression:

Linear regression relies on several assumptions:

### Linearity:
The relationship between the independent and dependent variables should be linear.

### Independence:
Residuals should be independent of each other. The residual for one observation should not depend on the residuals of other observations.

### Homoscedasticity:
The variance of the residuals should be constant across all levels of the independent variables.

### Normality:
The residuals should be normally distributed.

### No Multicollinearity (for Multiple Regression):
In multiple regression, independent variables should not be highly correlated with each other.

## Limitations of Linear Regression:

While linear regression is powerful, it has certain limitations:
- **Linearity Assumption**: If the relationship is not linear, the model won’t capture the true relationship.
- **Outliers**: Linear regression is sensitive to outliers, which can skew the results.
- **Multicollinearity**: In multiple regression, highly correlated independent variables can make the model unstable.
- **Overfitting**: Using too many variables can lead to overfitting, where the model fits the training data too closely but performs poorly on new data.

## Applications of Linear Regression:

Linear regression is widely used in various fields:
- **Economics**: Predicting GDP, inflation, or consumer behavior.
- **Marketing**: Predicting sales based on advertising spending.
- **Healthcare**: Estimating the impact of treatments or drug dosage on patient outcomes.
- **Finance**: Predicting stock prices or credit risk.

### 🧪 Model Evaluation: Understanding the Metrics

To know if our regression model is performing well, we use several key metrics:

1.  **MAE (Mean Absolute Error)**: The average 'distance' between the actual value and the prediction. Easy to interpret.
2.  **MSE (Mean Squared Error)**: Squares the errors before averaging. This penalizes large errors more heavily.
3.  **RMSE (Root Mean Squared Error)**: The square root of MSE. It brings the error unit back to the original scale (e.g., Dollars instead of Dollars²).
4.  **R² Score (Coefficient of Determination)**: Represents the proportion of variance for the dependent variable that's explained by the independent variables. Range: $(-\infty, 1]$.

After prediction a data scientist should always show how was the performance of the model using different evaluation metrics. One of the most important evaluation metric for linear regression is the coefficient of determination also known as R-squared value ($ R^2 $). The $ R^2 $ is basically a statistical measure of how close the data are to the fitted regression line, by "data" it means our independent variable data, the **engine-size** data. Usually, the larger the $R^2$ value is, the better the model fits the data.

```python
# R2 - How much of the variation in output is being incorporated by the input
# Is there an input justifying that variation in output
# 2 rows - which are identical
# but the output for the same are different
```

```python
# When my input ( TV spends ) is increasing OAPD and the output is also increasing OTPD
# OAD, my input increases, but my Output decreases, keeping everything else constant
# A high (close to 1) R2 score indicates that this isn't happenining
```

$ R^2 $ is defined as :

\begin{align}
r^2 = \frac{SS_{regression}}{SS_{total}}
\end{align}

\begin{align}
r^2 = 1 - \frac{SS_{error}}{SS_{total}}
\end{align}

Where:
 - $ SS_{regression} $ - "regression sum of squares"; quantifies how far the predicted value is from the mean
 - $ SS_{error} $ - "error sum of squares"; quantifies the variance of the target data points around the regression line
 - $ SS_{total} $ - "total sum of squares"; quantifies the variance of the target data points around the mean

\begin{align}
SS_{regression} = \sum_{i=1}^m(\hat y_i - \bar y)^2
\end{align}

\begin{align}
SS_{error} = \sum_{i=1}^m(y_i - \hat y)^2
\end{align}

\begin{align}
SS_{total} = \sum_{i=1}^m(y_i - \bar y)^2
\end{align}

```python
ss_total = np.sum((y_test - np.mean(y_test)) ** 2)  # Total sum of squares
ss_residual = np.sum((y_test - y_pred) ** 2)  # Residual sum of squares
r2 = 1 - (ss_residual / ss_total)
print("R-squared:", r2)
```

## Hypothesis Testing and p-values

Generally speaking, you start with a **null hypothesis** and an **alternative hypothesis** (that is opposite the null). Then, you check whether the data supports **rejecting the null hypothesis** or **failing to reject the null hypothesis**.

(Note that "failing to reject" the null is not the same as "accepting" the null hypothesis. The alternative hypothesis may indeed be true, except that you just don't have enough data to show that.)

As it relates to model coefficients, here is the conventional hypothesis test:
- **null hypothesis:** There is no relationship between TV ads and Sales (and thus $\beta_1$ equals zero)
- **alternative hypothesis:** There is a relationship between TV ads and Sales (and thus $\beta_1$ is not equal to zero)

How do we test this hypothesis? Intuitively, we reject the null (and thus believe the alternative) if the 95% confidence interval **does not include zero**. Conversely, the **p-value** represents the probability that the coefficient is actually zero:

```python
# Significant relationship between input and output or not
# p-value of an input feature <0.05 -> reject the null hypothesis, there is a significant relationship between input and output
# p-value -> >0.05 - there is no significant relationship
```

### Adjusted R-squared statistic

The Adjusted R-squared takes into account the number of independent variables used for predicting the target variable. In doing so, we can determine whether adding new variables to the model actually increases the model fit.

Let’s have a look at the formula for adjusted R-squared to better understand its working.

\begin{align}
AdjustedR^2 = {1-\frac{(1-r^2)(n-1)}{(n-k-1)}}
\end{align}

Where,

    n represents the number of data points in the dataset
    k represents the number of independent variables, and
    r represents the R-squared values determined by the model.


So, if R-squared does not increase significantly on the addition of a new independent variable, then the value of Adjusted R-squared will actually decrease.

On the other hand, if on adding the new independent variable we see a significant increase in R-squared value, then the Adjusted R-squared value will also increase.

```python
# Number of observations (n) and number of predictors (p)
n = len(y_test)  # Number of samples
p = X.shape[1]  # Number of independent variables (predictors)

# Calculate adjusted R-squared
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print("R-squared:", r2)
print("Adjusted R-squared:", adj_r2)
```

# Checking the The assumptions of Linear Regression
Those are crucial for ensuring that the model is valid and the results are reliable. You can check these assumptions using various diagnostic techniques.

## Linearity:<br>
The relationship between the independent and dependent variables should be linear.

**How to Check:**<br>
Plot the residuals (errors) against the predicted values. If the relationship is linear, the residuals should be randomly scattered around zero.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set modern theme
sns.set_theme(style="whitegrid", palette="muted")

# Residuals Calculation
residuals = y_test - y_pred

# Modern Residual Plot
plt.figure(figsize=(10, 6))
sns.residplot(x=y_pred, y=residuals, lowess=True,
             line_kws={'color': 'red', 'lw': 2, 'alpha': 0.8})

plt.xlabel('Predicted Sales', fontsize=12)
plt.ylabel('Residuals (Actual - Predicted)', fontsize=12)
plt.title('Residuals vs. Predicted Values (Linearity & Bias Check)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

- If the scatter plot shows no clear pattern and the points are randomly distributed, the linearity assumption holds.

## Independence of Errors:
The residuals should be independent of each other. This is particularly important for time-series data.

**How to Check:**<br>
Plot residuals against time or use statistical tests like the Durbin-Watson test.

```python
from statsmodels.stats.stattools import durbin_watson

# Durbin-Watson Test
dw = durbin_watson(residuals)
print(f"Durbin-Watson Statistic: {dw}")
```

- A value close to 2.0 indicates that there is no autocorrelation, while values less than 1.5 or greater than 2.5 suggest correlation.

## Homoscedasticity:
The variance of the residuals should be constant across all levels of the independent variables. In other words, the residuals should have the same spread for all predicted values.

**How to Check:**<br>
Plot the residuals against the predicted values. If the variance is constant, the residuals should be evenly spread around zero without a funnel shape.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_pred, y=residuals, color='#55a868', s=80, alpha=0.7)
plt.axhline(y=0, color='#c44e52', linestyle='--', linewidth=2)

plt.xlabel('Predicted Values', fontsize=12)
plt.ylabel('Residuals', fontsize=12)
plt.title('Homoscedasticity Check: Variance of Errors', fontsize=14, fontweight='bold')
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```

- If you see a funnel shape (i.e., increasing or decreasing spread of residuals as predicted values change), then there may be a problem with homoscedasticity.

## Normality of Residuals:
The residuals should be normally distributed. This can be checked by plotting a histogram or using a Q-Q plot (quantile-quantile plot).

**How to Check:**<br>
Use a histogram or Q-Q plot of the residuals.

```python
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 1. Distribution of Residuals
sns.histplot(residuals, kde=True, color="#4C72B0", ax=axes[0])
axes[0].set_title('Distribution of Error Terms', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Residual Value')

# 2. Probability Plot (Q-Q Plot)
stats.probplot(residuals, dist="norm", plot=axes[1])
axes[1].get_lines()[0].set_markerfacecolor('#C44E52')
axes[1].get_lines()[0].set_markersize(8)
axes[1].set_title('Normal Q-Q Plot', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

- If the residuals form a straight line on the Q-Q plot and the histogram is roughly bell-shaped, the normality assumption holds.

## Multicollinearity (For Multiple Linear Regression):
In multiple linear regression, the independent variables should not be highly correlated with each other.

**How to Check:**<br>
Use the Variance Inflation Factor (VIF) to check for multicollinearity. VIF values greater than 5 or 10 indicate potential multicollinearity.

# Polynomial Regression
Many relationships of interest, however, are not strictly linear. Often we want to model a non-linear relationship—for example, the relationship between the number of hours a student studies and the score she gets on the test. Intuitively, we can imagine there is a big difference in test scores between students who study for one hour compared to students who did not study at all. However, there is a much smaller difference in test scores between a student who studied for 99
hours and a student who studied for 100 hours. The effect one hour of studying has on a student’s test score decreases as the number of hours increases.

What if your data is actually more complex than a simple straight line? Surprisingly, you can actually use a linear model to fit nonlinear data. A simple way to do this is to add powers of each feature as new features, then train a linear model on this extended set of features. This technique is called Polynomial Regression.

```python
# X - X` ( original independent features - polynomial features )
# y -> y
```
