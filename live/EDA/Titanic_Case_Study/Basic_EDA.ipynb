# Let us Observe Age Distribution
----

did u know when you do analysis using a single variable(column) it is called Univariate Analysis ? 😕

```python
sns.distplot(df['Age'],hist=False)
plt.axvline(df['Age'].mean(),color='r')
plt.axvline(df['Age'].median(),color='g')
plt.show()
```

```python
# if your data is having extreme values eg.     most datapoints are between 15 and 44 but few are like 90 110  99 these are extreme values.

# if your data is having extreme values then you cannot use MEAN to replace Nulls, in this case median is suggested.
```
