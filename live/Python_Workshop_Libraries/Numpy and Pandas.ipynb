```python
#0-50 - Bronze
#51-100 - Silver
#101-150 - Gold 
#151-200 - Platinum
#201 - Diamond
```

```python
def cust_seg(tot_rev):
    if tot_rev<=50:
        return "Bronze Customer"
    elif tot_rev>50 and tot_rev<=100:
        return "Silver Customer"
    elif tot_rev>100 and tot_rev<=150:
        return "Gold Customer"
    elif tot_rev>150 and tot_rev<=200:
        return "Platinum Customer"
    else:
        return "Diamond Customer"
```

```python
tips["cs"] = tips.tot_rev.apply(cust_seg)
```

```python
tips.groupby("cs").count()
```

```python
def calc_rev(row):
    if row.gender == "Male":
        return (row.total_bill * row.noofcust) + row.tip
    else:
        return row.total_bill * row.noofcust

tips["tot_rev_new"] = tips.apply(calc_rev, axis=1)

#tips["cs"] = tips.tot_rev.apply(cust_seg)
```
