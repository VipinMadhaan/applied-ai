While it's possible to define your own functions for various tasks, the most common or useful ones have already been implemented by others. These are usually bundled together in libraries which each have their own purpose. For instance, the `random` library provides access to common random number generators. We can import the library to access these functions.

```python
import random
```

```python
random.randint(1, 10)  # generate a random integer between 1 and 10 by using the randint() function defined in the random library
```

```python
import numpy as np  # importing NumPy and calling it by the alias 'np' for ease of use / for the sake of convention
```

```python
np.sin(2 * np.pi)  # finding the sine of a value using the sin() function defined in the numpy library
```

In some cases, it is sufficient that you import only a particular module within a library. For instance, while importing the plotting library Matplotlib, most users conventionally import just the `pyplot` module.

```python
import matplotlib.pyplot as plt  # importing the pyplot module of matplotlib under the alias 'plt'

plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])  # plotting a number against its square using the plot() function in matplotlib.pyplot
```
