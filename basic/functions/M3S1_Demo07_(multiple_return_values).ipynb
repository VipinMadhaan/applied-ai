Functions can return multiple values

```python
def calculator(a, b):
    return a + b, a - b, a * b, a / b

x = calculator(5, 2)
print(x)
```

The output of functions returning multiple values may be **unpacked** for ease of use

```python
output1, output2, output3, output4 = calculator(5, 2)
print(output1)
print(output2)
print(output3)
print(output4)
```

Unneeded values can be assigned to the throwaway (placeholder) variable `_`

```python
add_out, _, _, div_out = calculator(5, 2)
print(add_out)
print(div_out)
print(_)  # the underscore variable name is only a convention for placeholders, it's still storing a value
```
