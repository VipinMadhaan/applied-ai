Here is a typical function

```python
def square_1(x):
    return x ** 2

print(square_1(5))
```

```python
type(square_1)
```

Lambda or anonymous functions are a compact way to write functions in a single line

```python
square_2 = lambda x: x ** 2

print(square_2(5))
```

```python
type(square_2)
```

```python
hello = lambda: print('Hello world!')  # this lambda function accepts no arguments and returns no values

hello()
```

```python
add = lambda x, y: x + y  # this lambda function accepts two arguments and returns one value

print(add(10, 15))
```

```python
square_and_root = lambda x: (x ** 2, x ** 0.5)  # this lambda function accepts one argument and returns two values

out1, out2 = square_and_root(25)
print(out1)
print(out2)
```

Lambda functions are useful when you don't plan on reusing a block of code too often; they don't even need to be assigned to a variable if used just once

```python
print((lambda x: x ** 2)(4))
```

We would still prefer a regular function defined using `def` if it requires detailed commentary and/or needs additional logic

```python
# Here, there is a conditional check for division by zero; it would be hard to incorporate this into a lambda function
def safe_divide(x, y):
    if y == 0: return 'Cannot divide by zero.'
    return x / y

print(safe_divide(4, 5))
print(safe_divide(5, 0))
```
