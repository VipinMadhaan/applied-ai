Let's define some sets

```python
set1 = {1, 2, 3, 4, 5, 6, 7, 8}
set2 = {0, 2, 4, 6, 8, 10}
set3 = {3, 4, 5}
set4 = {15, 16, 17}
```

Are the sets disjoint?

```python
set1.isdisjoint(set2)
```

```python
set1.isdisjoint(set4)
```

Let's find the union of two sets

```python
set1 | set2
```

```python
set1.union(set2)
```

Let's find the intersection of two sets

```python
set1 & set2
```

```python
set1.intersection(set2)
```

Let's find the difference between sets

```python
set1 - set2
```

```python
set1.difference(set2)
```

```python
set2 - set1
```

```python
set2.difference(set1)
```

Let's find the symmetric difference between sets

```python
set1 ^ set2
```

```python
set1.symmetric_difference(set2)
```

```python
set2 ^ set1
```

```python
set2.symmetric_difference(set1)
```
