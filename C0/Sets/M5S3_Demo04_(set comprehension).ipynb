Imagine you want to square all the elements in a range and create a set of these. One way to do this is by using a for loop.

```python
squares_set_1 = set()
for i in range(1, 10): squares_set_1.add(i ** 2)

squares_set_1
```

A set comprehension lets you create sets from sequences using a compact and readable block of code

```python
squares_set_2 = {x ** 2 for x in range(1, 10)}
squares_set_2
```

Set comprehensions can include conditionals

```python
squares_of_even = {x ** 2 for x in range(1, 10) if x % 2 == 0}
squares_of_even
```

```python
# Using if-else changes the syntax a bit
even_squares_odd_cubes = {x ** 2 if x % 2 == 0 else x ** 3 for x in range(1, 10)}
even_squares_odd_cubes  # seemingly, the order changes but it shouldn't be relied on
```

Sets automatically remove duplicates

```python
list1 = list(range(1, 10)) + list(range(2, 15))
list1
```

```python
set1 = {x for x in list1}
set1
```

```python
set(list1)
```
