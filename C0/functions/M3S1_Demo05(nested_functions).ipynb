Here's an example of a nested function

```python
def outer():
    print("I'm the outer function.")
    def inner():
        print("And I'm the inner function.")
    inner()

outer()
```

Let's explore what happens if we try to call `inner()` outside the `outer()`

```python
# inner()  # Uncommenting this will cause an error!
```

This will throw and error because `inner()` is local to the `outer()` and not accessible globally

Next, let's explore what happens if both `outer()` and `inner()` use a variable with the same name

```python
def outer():
    message = "Outer message"
    def inner():
        message = "Inner message"
        print(message)
    inner()
    print(message)

outer()
```

The `inner()` function prints its own message, which doesn't affect the variable in the `outer()`

 Now, what if we return `inner()` itself instead of calling it?  Let's try doing that using an example

```python
def num1(x):
    def num2(y):
        return x + y
    return num2

add10 = num1(10)  # returns num2 with x=10 remembered
print(type(add10))
print(add10(5))
```

Even though `num1` has finished, `num2` still remembers `x=10`

This still fails as Python treats `count` as local because of the assignment, and it hasn't been declared as `global`. Let's explicitly use the `global` keyword inside `outer` to modify the global `count`.

```python
count = 0
def outer():
    global count
    count += 1
    def inner():
        return count
    return inner()

outer()
```

```python
count
```

Thus, `outer()` successfully modifies the global `count`

Now, let's explore what happens if we use the `global` keyword inside `inner` to modify the global `count`

```python
count = 0
def outer():
    def inner():
        global count
        count += 1
        return count
    return inner()

outer()
```

```python
count
```

`inner()` also directly modifies the global `count` variable

What if we want to modify a variable in an enclosing scope that is not the global scope? This is where the `nonlocal` keyword is used. Let's go back to the scenario where `count` is inside `outer()`'s scope.

```python
def outer():
    count1 = 0
    def inner():
        nonlocal count1
        count1 += 1
        return count
    return inner()

outer()
```

The `nonlocal` keyword inside `inner` tells Python to modify the `count` variable in the nearest enclosing scope that is not the global scope

```python
# count1
```

It's important to remember that `nonlocal` cannot be used to modify variables in the global scope

As seen in above examples, using `nonlocal count` in `outer()` or `inner()` when `count` is a global variable results in an error. The `nonlocal` keyword is for variables in enclosing scopes, while the `global` keyword is for global scope.
