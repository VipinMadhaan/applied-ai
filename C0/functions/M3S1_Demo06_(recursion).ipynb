Recursion is when a function calls itself in order to solve a problem. A recursive function must have a recursive case when a function calls itself, and a base case where it does not.

Let's try to find the factorial of a number using recursion. A factorial $n!$ is given by:

$$n! = n \times (n-1) \times (n-2) \times \ldots \times 2 \times 1$$

For example, $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$.

Additionally, note that $0! = 1$.

```python
# Iterative approach to calculate the nth factorial
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i

    return result
```

```python
factorial_iterative(5)
```

```python
# Recursive approach to calculate the nth factorial
def factorial_recursive(n):
    if n == 0 or n == 1:  # base cases; function will terminate when these are encountered
        return 1
    else:
        return n * factorial_recursive(n - 1)  # recursive case; the function is calling itself with the argument n - 1
```

```python
factorial_recursive(5)
```

Generally speaking, iteration performs better than recursion, as the latter might involve a lot of repeated calculations, leading to a stack overflow and problems with memory. However, recursion tends to be more intuitive and elegant than iterative solutions, and are usually easier to understand. Additionally, there are certain complex problems which can only be solved using recursion. There are also algorithmic approaches such as the dynamic programming technique called memoisation which can assist with improving recursion to match or even outperform iteration.
