We an reuse a block of code by making a **function** out of it. We have previously encountered some in-built functions like `type()`, `input()`, and `print()`.

```python
def greeting(): print('Hello world!')
```

```python
def greeting():
    print('Hello world!')
```

A function can be **called** to execute its contents

```python
greeting()  # this is a function call
greeting()  # this is a function call
```

Variables defined within a function belong only to that function's **scope**

```python
def two_plus_two():
    x = 2
    y = 2
    print(x + y)

two_plus_two()
```

Functions can access variables defined outside the function (global variables) as long as they are assigned before the function call

```python
global_variable = 2

def two_plus_two():
    local_variable = 2
    print(global_variable + local_variable)

two_plus_two()
```

```python
print(global_variable)
```

Functions can **return** values using the `return` keyword. The `return` keyword is a jump statement.

```python
def two_plus_two():
    return(2 + 2)

answer = two_plus_two()
print(answer)
```

```python
def square_of_3():
    return 3 ** 2

print(square_of_3())
```

A `print()` statement just outputs the value being printed to the console, while a `return` statement allows you to retrieve the value from the function call

```python
def three_times_five():
    print(3 * 5)

answer = three_times_five()  # the internal print statement is executed but no value is returned to be stored in 'answer'
```

```python
answer
```

```python
type(answer)
```

```python
def four_times_six():
    return(4 * 6)

answer = four_times_six()  # the value of four times six is stored in 'answer'
```

```python
answer
```

```python
type(answer)
```

A function terminates upon returning a value

```python
def abc():
    return 'abc'
    print('This will not print.')

print(abc())
```

```python
def find_o():
    for ch in 'Hello world!':
        if ch == 'o':
            return 'o was found!'
        print(ch)

find_o()
```
