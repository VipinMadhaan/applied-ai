Here's an example of the **local scope** of a function

```python
def find_square(x):
    square = x * x
    return square

print(find_square(5))
```

```python
# print(square)  # this won't work as square is local to the scope of find_square()
```

Here is an example of the **global scope** of a program

```python
gravity = 9.81

def find_weight(mass):
    temp = mass * gravity  # the function can access any global variables
    return temp

print(find_weight(100))
```

```python
# print(temp)
```

```python
print(gravity)  # this will work as gravity is global to the scope of the program
```

What if we had two variables, one global and one local, with the same name?

```python
is_warm = 'no'

def check_if_warm(temperature):
    is_warm = False
    if temperature >= 25:
        is_warm = True

    return is_warm

print(check_if_warm(17))
```

```python
print(is_warm)  # once again the global variable is_warm remains unchanged
```

Under certain cases, though not recommended, you may want to modify a global variable inside a function. You can do this using the `global` keyword.

```python
counter = 0

def increment_counter():
    global counter  # mention that we are working with the global variable 'counter'
    counter += 1

increment_counter()
print(counter)  # here, the global variable 'counter' is being incremented from within the increment_counter() function
```

```python
counter = 0

def increment_counter():
    global counter  # the global 'counter' variable will be used throughout the scope of this function
    counter += 1
    counter = 100

increment_counter()
print(counter)
```
