Imagine you want to make a dictionary where every key corresponds to a value that is the square of that key. One way to do this is by using a for loop.

```python
squares_dict_1 = {}
for x in range(1, 10): squares_dict_1[x] = x ** 2

squares_dict_1
```

A dictionary comprehension lets you create dictionaries from sequences using a compact and readable block of code

```python
squares_dict_2 = {x: x ** 2 for x in range(1, 10)}
squares_dict_2
```

```python
my_dict = {'A': 1, 'B': 2, 'C': 3}
my_dict
```

```python
swapped_dict = {v: k for k, v in my_dict.items()}
swapped_dict
```

```python
words_list = ['alpaca', 'baboon', 'cheetah', 'duck']
word_lengths_dict = {word: len(word) for word in words_list}
word_lengths_dict
```

```python
list1 = ['apple', 'banana', 'cherry', 'durian']
list2 = ['tasty', 'tasty', 'alright', 'yuck']

zip(list1, list2)
```

```python
list(zip(list1, list2))
```

```python
fruits_taste_dict = {k: v for k, v in zip(list1, list2)}
fruits_taste_dict
```

Dictionary comprehensions can include conditionals

```python
squares_of_even = {x: x ** 2 for x in range(1, 10) if x % 2 == 0}
squares_of_even
```

```python
# Using if-else changes the syntax a bit
even_squares_odd_cubes = {x: (x ** 2 if x % 2 == 0 else x ** 3) for x in range(1, 10)}
even_squares_odd_cubes
```

Dictionary comprehensions can also be used to create nested dictionaries

```python
num_powers = {x: {y: x ** y for y in range(1, 4)} for x in range(1, 10)}
num_powers
```
