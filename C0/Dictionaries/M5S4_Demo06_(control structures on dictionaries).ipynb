Can you invert the keys and values of a dictionary? For instance, `{'Ashok': 1, 'Bina': 2, 'Charu': 3}` should invert to `{1: 'Ashok', 2: 'Bina', 3: 'Charu'}`.

````python
my_dict = {'Ashok': 1, 'Bina': 2, 'Charu': 3}
````

````python
inverted_dict = {}  # to store our inverted dictionary

for key, val in my_dict.items():
    inverted_dict[val] = key  # add a new key with the name of the old value and assign it to the old key

print(inverted_dict)
````

Given a list of words, can you group them into a dictionary where the keys are the lengths of the word and the values are lists of words of that length? For instance, the list `['apple', 'banana', 'cherry', 'durian', 'eldenberry', 'fig', 'guava']` should create the dictionary `{5: ['apple', 'guava'], 6: ['banana', 'cherry', 'durian'], 10: ['eldenberry'], 3: ['fig']}`.

````python
my_list = ['apple', 'banana', 'cherry', 'durian', 'eldenberry', 'fig', 'guava']
````

````python
lengths_dict = {}  # to store our output dictionary

for fruit in my_list:  # iterate through every element of the list
    length = len(fruit)
    if length in lengths_dict:  # if the length already exists as a key, append the word to the existing list
        lengths_dict[length].append(fruit)
    else:
        lengths_dict[length] = [fruit]  # if the length doesn't exist as a key, create a new key and assign it the value in a list

print(lengths_dict)
````

Can you count the occurrence of words in a sentence, and store this count in a dictionary? Ignore case while counting words.

````python
sentence = 'How much wood could a wood chuck chuck if a wood chuck could chuck wood'
````

````python
words = sentence.split()  # split the sentence into a list of words
word_count = {}

for word in words:
    word = word.lower()  # to ignore case
    if word in word_count:  # if the word exists as a key, increment its counter
        word_count[word] += 1
    else:  # if the word does not exist as a key, create the key and assign it the value 1
        word_count[word] = 1

print(word_count)
````

Given the following nested dictionary, can you print out the average scores across all subjects for each student, formatted to two decimal places?

```
students_scores = {
    'Ashok': {'Math': 85, 'English': 78, 'Science': 92},
    'Bina': {'Math': 90, 'English': 82, 'Science': 85},
    'Charu': {'Math': 88, 'English': 95, 'Science': 87},
    'Diana': {'Math': 76, 'English': 88, 'Science': 80},
}
```

````python
students_scores = {
    'Ashok': {'Math': 85, 'English': 78, 'Science': 92},
    'Bina': {'Math': 90, 'English': 82, 'Science': 85},
    'Charu': {'Math': 88, 'English': 95, 'Science': 87},
    'Diana': {'Math': 76, 'English': 88, 'Science': 80},
}
````

````python
# Iterate through the dictionary using dict.items()
for student, subjects in students_scores.items():
    total_score = 0
    subject_count = 0

    # Iterate over each student's subject scores
    for _, score in subjects.items():  # we do not require the subject name, so by convention we should use an underscore to indicate this
        total_score += score
        subject_count += 1

    # Calculate the average score for the current student
    average_score = total_score / subject_count
    print(f"{student}'s average score is: {average_score:.2f}")
````
