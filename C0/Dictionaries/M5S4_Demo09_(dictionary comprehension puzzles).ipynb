Given a dictionary with integers as values, can you create a new dictionary that contains only those key-value pairs where the value is odd?

````python
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
````

````python
odd_dict = {key: value for key, value in my_dict.items() if value % 2 != 0}
print(odd_dict)
````

Given a dictionary with city names as keys and their mean temperatures in degrees Celsius as values, can you create a new dictionary with the city names as keys and the temperature in degrees Fahrenheit as values?

**Note:** The temperature in degrees Fahrenheit ($F$) can be given by
$$F = C \times 9/5 + 32$$
where $C$ is the temperature in degrees Celsius.

````python
temps_celsius = {'Mumbai': 33, 'Bengaluru': 28, 'New Delhi': 32, 'Kolkata': 27}
````

````python
temps_fahrenheit = {city: temp * 9/5 + 32 for city, temp in temps_celsius.items()}
print(temps_fahrenheit)
````

Given a list containing dictionaries with student names, subjects, and scores, output a dictionary where every key is a subject and every value is another dictionary containing the average, minimum, and maximum scores in that subject across all students.

For example, consider the students as follows
```
students = [
    {'name': 'Ashok', 'subject': 'Maths', 'score': 87},
    {'name': 'Bob', 'subject': 'Maths', 'score': 91},
    {'name': 'Charu', 'subject': 'Maths', 'score': 76},
    {'name': 'Dawood', 'subject': 'English', 'score': 78},
    {'name': 'Esha', 'subject': 'English', 'score': 82},
    {'name': 'Falak', 'subject': 'English', 'score': 98},
    {'name': 'Girish', 'subject': 'Hindi', 'score': 91},
    {'name': 'Hasti', 'subject': 'Hindi', 'score': 85}
]
```
The output should be
```
{
  'Maths': {'average': 84.66666666666667, 'min': 76, 'max': 91},
  'English': {'average': 86.0, 'min': 78, 'max': 98},
  'Hindi': {'average': 88.0, 'min': 85,'max': 98}
}
```

````python
students = [
    {'name': 'Ashok', 'subject': 'Maths', 'score': 87},
    {'name': 'Bob', 'subject': 'Maths', 'score': 91},
    {'name': 'Charu', 'subject': 'Maths', 'score': 76},
    {'name': 'Dawood', 'subject': 'English', 'score': 78},
    {'name': 'Esha', 'subject': 'English', 'score': 82},
    {'name': 'Falak', 'subject': 'English', 'score': 98},
    {'name': 'Girish', 'subject': 'Hindi', 'score': 91},
    {'name': 'Hasti', 'subject': 'Hindi', 'score': 85}
]
````

We could use standard loops and conditionals to solve the problem

````python
# Use for loops to create lists for each subject that store the relevant scores
maths = []
english = []
hindi = []
for i in range(len(students)):
    current_item = students[i]
    if current_item['subject'] == 'Maths': maths.append(current_item['score'])
    elif current_item['subject'] == 'English': english.append(current_item['score'])
    else : hindi.append(current_item['score'])
````

````python
# Create a new dictionary with the required values
output_dict = {}
output_dict['Maths'] = {'average': sum(maths) / len(maths), 'min': min(maths), 'max': max(maths)}
output_dict['English'] = {'average': sum(english) / len(english), 'min': min(english), 'max': max(english)}
output_dict['Hindi'] = {'average': sum(hindi) / len(hindi), 'min': min(hindi), 'max': max(hindi)}
````

````python
output_dict
````

We can also do this using comprehensions

````python
# Extracting the unique subject names in the original data
subject_names = []
for i in range(len(students)):
    current_item = students[i]
    if len(subject_names) == 0: subject_names.append(current_item['subject'])
    else:
        if current_item['subject'] not in subject_names: subject_names.append(current_item['subject'])
````

````python
subject_names
````

````python
# Create an intermediate dictionary to hold the relevant scores
temp_dict = {subject: [students[i]['score'] for i in range(len(students)) if students[i]['subject'] == subject] for subject in subject_names}
````

````python
temp_dict
````

````python
# Create a new dictionary with the required values
output_dict = {subject: {'average': sum(temp_dict[subject]) / len(temp_dict[subject]),
                         'min': min(temp_dict[subject]),
                         'max': max(temp_dict[subject])} for subject in subject_names}
````
