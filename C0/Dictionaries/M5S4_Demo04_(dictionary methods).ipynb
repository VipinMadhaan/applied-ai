Dictionaries have various methods associated with them

```python
my_dict = {'student': 'Ashok Kumar', 'age': 16, 'sports': ['football', 'cricket', 'basketball']}
my_dict
```

We can retrieve values from keys safely using `.get()`. This function returns nothing when the key does not exist, unlike using subscripts which will throw an error.

```python
my_dict.get('student')
```

```python
my_dict.get('height')  # returns nothing
```

```python
my_dict
```

We can retrieve all the keys and/or values within our dictionary

```python
my_dict.keys()  # retrieve all keys as a dict_keys object
```

```python
# my_dict.keys()[0]  # a dict_keys object behaves more like a set than a list, and is not subscriptable
```

```python
list_of_keys = list(my_dict.keys())
list_of_keys
```

```python
list_of_keys[0]
```

```python
my_dict.values()  # retrieve all values as a dict_values object
```

```python
# my_dict.values()[0]
```

```python
list_of_values = list(my_dict.values())
list_of_values
```

```python
list_of_values[0]
```

```python
my_dict.items()  # retrieve all key-value pairs as a dict_items object
```

```python
# my_dict.items()[0]
```

```python
list_of_items = list(my_dict.items())
list_of_items
```

```python
list_of_items[0]  # 'pairs' are a type of tuple in Python (2-tuple)
```

As dictionaries are mutable, the methods which modify dictionaries act in place, i.e. they do not require to be reassigned to a variable for the changes to act

We can add elements to our dictionary

```python
my_dict.update({'roll number': 12})
my_dict
```

We can remove elements from our dictionary

```python
popped_value = my_dict.pop('student')  # uses the key to remove the key-value pair but also returns the value
popped_value
```

```python
my_dict
```

```python
popped_item = my_dict.popitem()  # removes the last inserted key-value pair and returns it
popped_item
```

```python
my_dict
```

```python
my_dict.clear()  # empties the dictionary
my_dict
```
