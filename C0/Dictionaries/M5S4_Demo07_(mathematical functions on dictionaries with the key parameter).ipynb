Consider the following dictionary of names and scores. How would you find the key with the maximum score?

```python
scores = {
    'Ashok': 83,
    'Bina': 97,
    'Caruna': 70,
    'David': 85,
}
```

What would happen if we were to use the `max()` function on this dictionary?

```python
max(scores)
```

This isn't what we intended, as we want to get the highest score, not the string that is considered the greatest (because 'David' starts with 'D'). We can instead try:

```python
max(scores, key = scores.get)
```

This is the same as doing:

```python
max(scores, key = lambda person: scores[person])
```

Here, we are sending each key as an argument to the lambda function. We are then querying the dictionary with that key. The `max()` function will compare the values obtained from this querying operation for all keys, and return the largest.

Similarly, we can find the minimum value in the dictionary by score using one of the following methods:

```python
min(scores, key = scores.get)
```

```python
min(scores, key = lambda person: scores[person])
```

We can also sort the dictionary by the scores in ascending order

```python
sorted(scores, key = scores.get)  # returns a list of keys sorted in ascending order
```

```python
sorted(scores, key = lambda person: scores[person])
```

Or we could sort the dictionary by the scores in descending order

```python
sorted(scores, key = scores.get, reverse = True)
```

```python
sorted(scores, key = lambda person: scores[person], reverse = True)
```

What if we had a slightly more complex dictionary? The same logic would apply.

```python
people = {
    'Arvind': {'age': 30, 'height': 175},
    'Bob': {'age': 25, 'height': 180},
    'Charu': {'age': 27, 'height': 165}
}
```

```python
max(people)  # here, we are just comparing the keys
```

```python
# Here, the lambda function will retrieve the corresponding age for each key
# The max() function will then compare these retrieved ages and find the maximum
max(people, key = lambda person: people[person]['age'])
```

```python
max(people, key = lambda person: people[person]['height'])
```

```python
min(people)
```

```python
min(people, key = lambda person: people[person]['age'])
```

```python
min(people, key = lambda person: people[person]['height'])
```

```python
sorted(people)  # returns a list
```

```python
sorted(people, key = lambda person: people[person]['age'])
```

```python
sorted(people, key = lambda person: people[person]['age'], reverse = True)
```

```python
sorted(people, key = lambda person: people[person]['height'])
```

```python
sorted(people, key = lambda person: people[person]['height'], reverse = True)
```
