Let's imagine we work in a library where there's a dictionary called `library` that stores information about each book in the library. Each book is identified by an ISBN number and contains details such as title, author, price, and whether the book is in stock or not.

```python
library = {
    '978-0140449136': {
        'title': 'The White Tiger',
        'author': 'Aravind Adiga',
        'price': 499,
        'in_stock': True
    },
    '978-0679783272': {
        'title': 'Pride and Prejudice',
        'author': 'Jane Austen',
        'price': 299,
        'in_stock': False
    },
    '978-0743273565': {
        'title': 'The Great Gatsby',
        'author': 'F. Scott Fitzgerald',
        'price': 399,
        'in_stock': True
    },
    '978-0156012195': {
        'title': 'The Alchemist',
        'author': 'Paulo Coelho',
        'price': 399,
        'in_stock': True
    },
}
```

How can we safely retrieve a book by ISBN number, so that an error isn't caused if the number doesn't exist in our library?

```python
def get_book_by_isbn(isbn):
    book = library.get(isbn)  # is None if no book is found
    if book:
        print(f'Book found: {book["title"]} by {book["author"]}, Price: ₹{book["price"]}')
    else:
        print('Book not found in the library.')
```

```python
get_book_by_isbn('978-0743273565')
```

```python
get_book_by_isbn('0000000000')
```

Let's retrieve a list of all ISBNs available in the library

```python
list(library.keys())
```

Let's print out information on every book available in the library

```python
print('Books in the library:')
for book in library.values():
    print(f'{book["title"]} by {book["author"]}, Price: ₹{book["price"]}, In Stock: {book["in_stock"]}')
```

We also want to display the ISBNs with the rest of the information. How can we achieve this?

```python
print('Books in the library:')
for isbn, book in library.items():
    print(f'ISBN {isbn}: {book["title"]} by {book["author"]}, Price: ₹{book["price"]}, In Stock: {book["in_stock"]}')
```

Someone returned Pride and Prejudice (ISBN 978-0679783272) back to our shelves. How can we update our dictionary to reflect this?

```python
library['978-0679783272'].update({'in_stock': True})
print(library['978-0679783272'])
```

Our library's copy of The Great Gatsby (ISBN 978-0743273565) was lost. How can we update our dictionary to reflect this? Also print out that the book was lost.

```python
removed_book = library.pop('978-0156012195')
print(f'Removed book: {removed_book["title"]} by {removed_book["author"]}')

library
```
