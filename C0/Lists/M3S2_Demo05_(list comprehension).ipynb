Imagine you want to square all the elements of a list and create a new list out of this. One way to do this is by using a for loop.

```python
my_list = [1, 2, 3, 4, 5, 6, 7]

squared_list = []
for ele in my_list: squared_list.append(ele ** 2)

squared_list
```

A list comprehension lets you create lists from sequences using a compact and readable block of code

```python
squared_list = [x ** 2 for x in my_list]
squared_list
```

List comprehensions can include conditionals

```python
my_string = 'InfOrmAtioN'

new_list = [ch.upper() for ch in my_string if ch.islower()]
new_list
```

```python
# Using if-else changes the syntax a bit
new_list_2 = [ch.upper() if ch.islower() else ch for ch in my_string]
new_list_2
```

List comprehensions can also be used to create nested lists

```python
table = [[x * y for y in range(1, 6)] for x in range(1, 6)]
table
```
