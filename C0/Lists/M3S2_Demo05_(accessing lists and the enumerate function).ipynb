We have a 2D list containing stock pricing data for various companies. Let's work through querying this list to extract some useful data.

```python
# Rows represent companies; columns represent days
stock_prices = [
    [150, 152, 148, 153, 155],  # company A
    [78, 80, 7, 76, 78],  # company B
    [210, 215, 220, 219, 222]  # company C
]
```

Let's retrieve the stock price for company B on day 3

```python
stock_prices[1][2]
```

Let's retrieve all prices for a company C as a list

```python
stock_prices[2]
```

Let's retrieve the prices for every company on day 5

```python
[stock_prices[i][4] for i in range(len(stock_prices))]
```

**Note:** The `enumerate()` function lets us iterate through a list and use both its index as well as its value at that index

Let's print out the average stock prices for every company, formatted to two decimal places

```python
for i, company in enumerate(stock_prices):
    avg_price = sum(company) / len(company)
    print(f'Average stock price for Company {chr(65 + i)}: {avg_price:.2f}')
```

Let's print out the highest and lowest prices in the week for each company

```python
for i, company in enumerate(stock_prices):
    highest_price = max(company)
    lowest_price = min(company)
    print(f'Company {chr(65 + i)}: Highest price = {highest_price}, Lowest price = {lowest_price}')  # using chr() to print the company name as an alphabet
```

Let's print out the average stock price across every company on each day, formatted to two decimals

```python
n_days = len(stock_prices[0])  # get the length of a week
for day in range(n_days):
    daily_prices = [stock_prices[i][day] for i in range(len(stock_prices))]
    daily_avg = sum(daily_prices) / len(daily_prices)
    print(f'Average stock price across companies on day {day + 1}: {daily_avg:.2f}')
```

Let's print out the prices for each company from day 3 to day 5 as a list

```python
for i, company in enumerate(stock_prices):
    subset_prices = company[2:]
    print(f'Prices for Company {chr(65 + i)} from day 3 to day 5: {subset_prices}')
```
