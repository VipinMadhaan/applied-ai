Let's make a to-do list of tasks which we might need to do on our developer job

```python
todo_list = ['Design user interface', 'Write API documentation', 'Fix login bug']
```

We've realised we would be the ones deploying the code to production. Let's add this as a task `'Deploy to production'` to our list

```python
todo_list.append('Deploy to production')
todo_list
```

We've received a new task `'Meet with stakeholders'` which has a higher priority, as it involves high-profile individuals. We want to add it to the top of our list.

```python
todo_list.insert(0, 'Meet with stakeholders')
todo_list
```

We've been given a bunch of other tasks `'Review security protocols'` and `'Conduct performance testing'` because our coworker is ill. They aren't very high priority so we can add them to the end of our list. Can you do this in one statement?

```python
todo_list.extend(['Review security protocols', 'Conduct performance testing'])
todo_list
```

We've managed to quickly resolve the `'Fix login bug'` task, let's remove it from our list

```python
todo_list.remove('Fix login bug')
todo_list
```

Our colleague told us that they finished the third task on our to-do list, but you're not sure what this task was called. Our colleague is in a meeting now, so we can't ask them. How do we remove the third task using its position? How can we find out the name of this task?

```python
finished_task = todo_list.pop(2)
print('Finished task:', finished_task)
print(todo_list)
```

After all these additions and removals, we're unsure as to which task our `'Deploy to production'` task is in terms of priority. How can we find its index?

```python
index_of_task = todo_list.index('Deploy to production')
print("Index of 'Deploy to production':", index_of_task)
```

On a whim, our boss has asked us to ignore the original priorities of our tasks, and to do them alphabetically! How can we accommodate this?

```python
todo_list.sort()
todo_list
```

Wait! They meant we should do the tasks in reverse alphabetical order. How can we reverse our alphabetically sorted list?

```python
todo_list.reverse()
todo_list
```

The day is done and, miraculously, we managed to finish all of our tasks. Let's clear out our to-do list so we can use it again tomorrow.

```python
todo_list.clear()
todo_list
```
