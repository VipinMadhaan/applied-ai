Lists have various methods associated with them

```python
my_list = [1, 2, 3, 4, 5, 6, 7]
my_list
```

We can find the index of an element within the list

```python
my_list.index(3)
```

```python
# my_list.index(8)  # this element doesn't exist in the list
```

```python
my_list.index(max(my_list))
```

As lists are **mutable,** the methods which modify lists act inplace, i.e. the modified list does not need to be reassigned to a variable for the changes to apply

We can add elements to a list

```python
my_list.append(8)
my_list
```

```python
temp = my_list.append(9)
temp
```

```python
my_list
```

```python
my_list.insert(4, 'inserted value')
my_list
```

```python
my_list.extend(['more', 'stuff'])
my_list
```

We can remove elements from a list

```python
popped_value = my_list.pop()  # the value is removed from the list and also returned
popped_value
```

```python
my_list
```

```python
my_list.remove(6)
my_list
```

```python
my_list.clear()
my_list
```

We can reverse a list

```python
new_list = [4, 3, 8, 9, 1, 7, 10]
new_list.reverse()

new_list
```

We can sort a list in ascending or descending order

```python
new_list.sort()
new_list
```

```python
new_list.sort(reverse = True)
new_list
```
