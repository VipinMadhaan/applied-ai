The `zip()` function lets you combine two or more iterable sequences into a single iterable sequence. This is valuable when you want to iterate over multiple iterable sequences at the same time.

```python
lst1 = ['a', 'b', 'c', 'd', 'e']
lst2 = [1, 2, 3, 4, 5]
```

```python
zipped_lists = zip(lst1, lst2)  # zipping two lists
zipped_lists
```

```python
type(zipped_lists)
```

```python
list(zipped_lists)  # creating a list out of the zip object; the internal elements in parantheses are called tuples
```

```python
list(zipped_lists)  # the zip object has already been exhausted; it no longer contains any values
```

Zip works with more than two iterable sequences. Let's try this out with a concrete example.

```python
names = ['Ashok', 'Bina', 'Charu', 'David']
ages = [32, 26, 30, 27]
cities = ['Mumbai', 'New Delhi', 'Bengaluru', 'Chennai']
```

```python
zipped_lists = zip(names, ages, cities)
```

```python
for name, age, city in zipped_lists:
    print(f'{name} is {age} years old and lives in {city}.')
```

```python
# This does nothing because the zip object has already been exhausted!
for name, age, city in zipped_lists:
    print(f'{name} is {age} years old and lives in {city}.')
```
