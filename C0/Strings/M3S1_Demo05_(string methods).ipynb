Methods are a set of instructions that perform a task, like a function. However, methods are associated with an object while functions are not. When we create a string, we are creating an object of the class `str`. Let's initialise a string and run some pre-built methods on it.

```python
message = 'python programming is fun!'
message
```

We can change the case of a string

```python
message.upper()
```

String methods do not act in place, i.e. they do not change the object unless the variable is reassigned. This is because strings are **immutable** data structures.

```python
message
```

```python
message_uppercase = message.upper()
message_uppercase
```

```python
message.lower()
```

```python
message.capitalize()
```

```python
message.title()
```

```python
random_string = 'aBcDeFg'
random_string.swapcase()
```

We can check if strings contain digits or alphabets

```python
num_string = '123'
num_string.isnumeric()
```

We can find the **index** of a character in a string. An index refers to a position or location within a sequence, allowing you to access individual elements in that sequence.

```python
message.index('y')  # index counting starts from 0 in Python
```

```python
message.index('thon')
```

```python
# message.index('X')  # .index() will throw an error if the substring isn't in the string
```

```python
message.find('y')
```

```python
message.find('thon')
```

```python
message.find('X')  # .find() will return -1 if the substring isn't in the string
```

```python
message.count('g')
```

We can replace characters or substrings with other strings

```python
message.replace('m', 'X')
```

```python
message.replace('fun', 'interesting')
```
