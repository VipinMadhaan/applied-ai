Strings, being iterable sequences, can be traversed through using loops. This can have various applications. For instance, we could count the number of vowels in a string.

```python
my_string = 'my cat is brown and golden'  # statement to check
vowels = 'aeiouAEIOU'  # a string containing vowels, in both, lower and upper case
counter = 0

for ch in my_string:  # loop through each character in the string
    if ch in vowels:  # check if the character is in our string containing vowels
        counter += 1  # increment our counter if it is

print(f'There are {counter} vowels in the phrase "{my_string}"')
```

Imagine that we are used to a certain password but it unfortunately contains very common words. We decide to encrypt the password using the Caesar Cipher.

The Caesar Cipher accepts a 'shift' value, and shifts all letters to the right depending on the shift value. When the character goes past 'z', we want to return to 'a' (wraparound). If the character isn't a letter, we don't change it. How can we code something like this?

For this, we could use the following functions:

*   `ord()` to obtain the ASCII value of a character
*   `chr()` to convert an ASCII value back to a character

Let's experiment with these functions a bit before writing our encryption program.


```python
def caesar_cipher(password, shift):
    result = ''  # to store the encrypted password

    # Loop through each chacter in the string
    for ch in password:

        # Case for when the character is in upper case
        if ch.isupper():
            # Find the position of the letter in the alphabet (0-25), apply shift, and handle wraparound
            shifted = (ord(ch) - ord('A') + shift) % 26
            new_ch = chr(shifted + ord('A'))  # convert back to a letter
            result += new_ch  # add this to our new string to be output

        # Case for when the character is in lower case
        elif ch.islower():
            # Find the position of the letter in the alphabet (0-25), apply shift, and handle wraparound
            shifted = (ord(ch) - ord('a') + shift) % 26
            new_ch = chr(shifted + ord('a'))  # Convert back to a letter
            result += new_ch

        # Case for when the character is not a letter - we leave the character unchanged
        else:
            result += ch

    return result
```

```python
my_password = input('My password: ')
encrypted_password = caesar_cipher(my_password, 6)
print('Encrypted password:', encrypted_password)
```
