Let's write a program to print out a restaurant bill. The bill must acount for the tax rate as well as a tip rate. The final bill should look like this. Assume there are whitespaces one tab long after each colon.

```
Restaurant Bill
----------------
Subtotal:       ₹512.75
Tax (7.5%):     ₹38.46
Tip (18%):      ₹92.30
----------------
Total:          ₹643.50
```

````python
subtotal = 512.75
tax_rate = 0.075
tip_rate = 0.18

tax = subtotal * tax_rate
tip = subtotal * tip_rate
total = subtotal + tax + tip
````

````python
print(f'Restaurant Bill\n----------------\nSubtotal:\t₹{subtotal}\nTax (7.5%):\t₹{tax}\nTip (18%):\t₹{tip}\n----------------\nTotal:\t\t₹{total}')
````

That doesn't look too good. Perhaps some float formatting could help us out. Let's round everything up to two digits.

````python
print(f'Restaurant Bill\n----------------\nSubtotal:\t₹{subtotal:.2f}\nTax (7.5%):\t₹{tax:.2f}\nTip (18%):\t₹{tip:.2f}\n----------------\nTotal:\t\t₹{total:.2f}')
````
