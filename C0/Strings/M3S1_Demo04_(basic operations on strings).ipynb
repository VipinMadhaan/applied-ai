We can finding the length of a string using `len()`

```python
greeting = 'Hello world!'
print(len(greeting))
```

We can find the ASCII value of a character using `ord()`. ASCII (American Standard Code for Information Interchange) is a standard that assigns unique numeric codes to characters, which include letters, digits, punctuation marks, control characters, and other symbols. These numeric values allow computers to store and manipulate characters as numbers.

```python
ord('A')
```

```python
ord('B')
```

When comparing strings, ASCII values are used for comparison

```python
'A' > 'B'
```

```python
'Apple' > 'Banana'  # comparison will compare the ASCII values of the first dissimilar characters in a string ('A' versus 'B' in this case)
```

```python
'o' > 'i'
```

```python
'Python' > 'Pything'  # comparison will compare the ASCII values of the first dissimilar characters in a string ('o' versus 'i' in this case)
```

```python
str1 = 'Python'
str2 = 'programming'

str1 + str2  # concatenating a string
```

```python
str1 + ' ' + str2 + ' is fun!'
```
