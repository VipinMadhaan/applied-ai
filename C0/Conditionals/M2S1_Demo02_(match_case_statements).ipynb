The `match-case` statement can be used as an alternative to `if-elif-else` code blocks. For example, consider the following code block:

```python
code = int(input())
if code == 200:
    print('OK')
elif code == 404:
    print('Not found')
elif code == 500:
    print('Internal server error')
else:
    print('Unknown error')
```

The following is way offer the same functionality using `match-case`:

```python
# This block of code will accept an HTTP status code and reply with a message
code = int(input())
match code:
    case 200:
        print('OK')
    case 404:
        print('Not found')
    case 500:
        print('Internal server error')
    case _:  # wildcard which covers all other cases
        print('Unknown error')
```

There are certain cases where `match-case` offers a cleaner way to do things than `if-elif-else`. For instance, consider the following program which tells us where a point lies on the coordinate axis:

```python
x = int(input('Input x-coordinate ... '))
y = int(input('Input y-coordinate ... '))

if x == 0 and y == 0:
    print('The point lies on the origin')
elif x == 0 and y != 0:
    print('The point lies on the y-axis at y =', y)
elif x != 0 and y == 0:
    print('The point lies on the x-axis at x =', x)
else:
    print('The point lies at x = ' + str(x) + ' and y = ' + str(y))
```

We could use `match-case` to simplify this and make it more intuitive to read

```python
x = int(input('Input x-coordinate ... '))
y = int(input('Input y-coordinate ... '))

match x, y:
    case 0, 0:
        print('The point lies on the origin')
    case 0, y:
        print('The point lies on the y-axis at y =', y)
    case x, 0:
        print('The point lies on the x-axis at x =', x)
    case _:
        print('The point lies at x = ' + str(x) + ' and y = ' + str(y))
```

You can add **guard clauses** which are essentially conditions in `match-case`

```python
num = int(input())
match num:
    case x if x < 0:
        print('Negative number')
    case x if x == 0:
        print('Zero')
    case x if x > 0:
        print('Positive number')
    case _:
        print('Not a number')
```

You can use the `|` operator to specify multiple accepted match patterns

```python
num = int(input('Enter a number between 2 and 9 ... '))
match num:
    case 2 | 3 | 5 | 7:
        print('The number is prime')
    case 4 | 6 | 8 | 9:
        print('The number is not prime')
    case _:
        print('Not a number or the number is not between 2 and 9')
```
