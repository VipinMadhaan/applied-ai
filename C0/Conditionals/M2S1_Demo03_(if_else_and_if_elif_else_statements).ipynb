We may want our code to perform a task when an `if` statement does not execute. An `else` statement executes if the preceding `if` statement does not.

```python
x = 7
if x > 6: print('This will print if True.')
else: print('Or else this will print.')
```

```python
x = 5
if x > 6: print('This will print if True.')
else: print('Or else this will print.')
```

```python
if True: print('This will always print.')
else: print('This will never print')
```

```python
if x > 6: x += 1
else: x -= 1
print(x)
```

```python
marks = 80
if marks >= 90: print('Grade A')
if marks >= 80 and marks < 90: print('Grade B')
if marks >= 70 and marks < 80: print('Grade C')
if marks >= 60 and marks < 70: print('Grade D')  # this is the only 'if' statement that the 'else' will consider
else: print('Fail')
```

If we want to chain conditionals one-after-another, we can use `elif`, which stands for 'else-if'

```python
if marks >= 90: print('Grade A')
elif marks >= 80 and marks < 90: print('Grade B')
elif marks >= 70 and marks < 80: print('Grade C')
elif marks >= 60 and marks < 70: print('Grade D')  # this is the only 'elif' statement that the 'else' will consider
else: print('Fail')
```

Let's imagine we are designing a system to help colourblind people understand what the colour being displayed by a traffic light represents. We can create a simple block of code where the user provides an input colour and the program prints out what the colour represents.

```python
colour = input()
if colour == 'red':
    print('Stop!')
elif colour == 'yellow':
    print('Wait.')
elif colour == 'green':
    print('Go!')
else:
    print('<invalid colour>')
```
