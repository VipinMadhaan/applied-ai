You can **nest** conditionals to perform more complex decision-making logic. In programming, nesting is the practice of placing one programming construct within another.

Here is a block of code which asks the user for their age and then tells them if they are an adult or not. If they are an adult, it further tells them if they are allowed to drink alcohol in Maharashtra or not.

```python
age = int(input())
if age > 18:
    print('You are an adult.')
    if age >= 25:  # this if statement is nested within the previous
        print('You are allowed to drink any alcohol in Maharashtra.')  # note the additional indentation here
    elif age >= 21:
        print('You are allowed to drink beer in Maharashtra.')
    else:
        print('You are not allowed to drink alcohol in Maharashtra.')
else:
    print('You are a minor.')
```

Here is a block of code which takes the credit score and income of the user as an input. It then determines if the user is eligible for a credit card or not.

```python
credit_score = int(input('Input your credit score: '))
income = int(input('Input your monthly income: '))
if credit_score >= 700:
    if income >= 50000:
        print('Credit card appoved!')
    else:
        print('Insufficient income.')
else:
    print('Low credit score.')
```
