Let's implement the logic for a simple two-elevator system (let's call them Lift A and Lift B). The idea is to choose the right lift depending on the floor from which they are being hailed.

Let's define our variables here:
- The calling floor
- The floor in which Lift A is at rest or the floor to which it is moving
- The floor in which Lift B is at rest or the floor to which it is moving

```python
# Input variables
floor_call = 11
lift_A = 3
lift_B = 19

# Logic for choosing the right lift for the calling floor
dist_A = abs(lift_A - floor_call)
dist_B = abs(lift_B - floor_call)

if dist_A < dist_B: print('Lift A is chosen')
else: print('Lift B is chosen')  # this includes the case where the two distances are equal
```

Let's improve the code to handle two calling floors.

Let's define our variables here:
- The calling floors, say X and Y
- The floor in which Lift A is at rest or the floor to which it is moving
- The floor in which Lift B is at rest or the floor to which it is moving

```python
# Input variables
floor_call_X = 2
floor_call_Y = 8
lift_A = 5
lift_B = 7

# Logic for choosing the right lift for the two calling floors
dist_AX = abs(lift_A - floor_call_X)
dist_BX = abs(lift_B - floor_call_X)
dist_AY = abs(lift_A - floor_call_Y)
dist_BY = abs(lift_B - floor_call_Y)

# Total distances of each lift from each floor in the two possible cases
dist_AX_BY = dist_AX + dist_BY  # Lift A goes to Floor X and Lift B goes to Floor Y
dist_AY_BX = dist_AY + dist_BX  # Lift A goes to Floor Y and Lift B goes to Floor X

if dist_AX_BY < dist_AY_BX: print('Lift A is chosen for Floor X and Lift B is chose for Floor Y')
else: print('Lift B is chosen for Floor X and Lift A is chosen for Floor Y')  # this includes the case where the two distances are equal
```

Let's take another example. This time, let's consider the logic for the discount amount at a store.

Suppose the conditions are as follows:
- If you buy for less than INR 2500, then no discount
- If you buy for INR 2500 or more, then 5% discount
- If you buy for INR 5000 or more, then 10% discount

The only variable here is the buying amount

```python
amount = 1500

if amount < 2500: print('no discount')
else:
    if amount < 5000: print('5% discount')
    else: print('10% discount')
```

We can do this using logical operators as well

```python
amount = 6500

if amount < 2500: print('no discount')
elif amount >= 2500 and amount < 5000: print('5% discount')
else: print('10% discount')
```

```python
amount = 10000

# Storing conditions
c1 = amount < 2500
c2 = amount >= 2500 and amount < 5000

if c1: print('no discount')
elif c2: print('5% discount')
else: print('10% discount')
```
