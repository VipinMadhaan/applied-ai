We can **unpack** the elements of a tuple

```python
my_tuple = (1, ['a', 'b', 'c'], 'Python', (1, 2, 3))
my_tuple
```

```python
ele1, ele2, ele3, ele4 = my_tuple
ele1
```

```python
ele2
```

```python
ele3
```

```python
ele4
```

We can use an asterisk (`*`) while unpacking to assign the remaining elements of a sequence to a variable as a list. This is helpful when we don't know the number of elements or do not want to access them right away.

```python
ele1, *rest = my_tuple
ele1
```

```python
rest
```

```python
ele1, *rest, ele2 = my_tuple
ele1
```

```python
rest
```

```python
ele2
```

We can use an underscore (`_`) while unpacking to assign throwaway (unnecessary) elements

```python
ele1, _, _, ele4 = my_tuple
ele1
```

```python
ele4
```

```python
_  # still stores the last assigned value, it's only convention to use an underscore
```

When functions return multiple values, they return them as a tuple which can then be unpacked

```python
def calculator(a, b): return a + b, a - b, a * b, a / b

x = calculator(5, 2)
print(x)
```

```python
output1, output2, output3, output4 = calculator(5, 2)
print(output1)
print(output2)
print(output3)
print(output4)
```
